```python

def jump(nums):

    n = len(nums)
    if n <= 1:
        return 0
    
    jumps = 0           # Number of jumps made
    current_end = 0     # End of current jump range
    farthest = 0        # Farthest position we can reach
    
    # We don't need to jump from the last position
    for i in range(n - 1):
        # Update the farthest position we can reach
        farthest = max(farthest, i + nums[i])
        
        # If we've reached the end of current jump range
        if i == current_end:
            jumps += 1
            current_end = farthest
            
            # Early exit if we can already reach the end
            if current_end >= n - 1:
                break
    
    return jumps
```

```python
def canJump(nums):

    max_reach = 0
    length = len(nums)
    for i in range(length):
        if i > max_reach:
            return False
        
        max_reach = max(max_reach, i + nums[i])
        if max_reach >= length - 1:
            return True
        
    return True
```

```python
print(canJump([3,2,1,0,4]))
print(canJump([2,3,1,1,4]))
```

```python
def canCompleteCircuit(gas, cost):
    """
    Find the starting gas station index to complete the circular route.
    
    Args:
        gas: List of gas amounts at each station
        cost: List of costs to travel to next station
    
    Returns:
        Starting station index, or -1 if impossible
    """
    n = len(gas)
    total_gas = 0
    total_cost = 0
    current_gas = 0
    start = 0
    
    for i in range(n):
        total_gas += gas[i]
        total_cost += cost[i]
        current_gas += gas[i] - cost[i]
        
        # If we can't reach next station from current start
        if current_gas < 0:
            # Try starting from next stationï¼Œbefore i are all reachable, so do not need to be mind
            # doesn't clear total_gas and total_cost, so still use this to check
            start = i + 1
            current_gas = 0
    
    # If total gas is less than total cost, impossible
    if total_gas < total_cost:
        return -1
    
    return start
```

```python
def eraseOverlapIntervals(intervals):

    if not intervals:
        return 0
    
    # Sort by end time (greedy: prefer intervals that end earlier)
    intervals.sort(key=lambda x: x[1])
    
    removed = 0
    prev_end = intervals[0][1]
    
    for i in range(1, len(intervals)):
        start, end = intervals[i]
        
        if start < prev_end:
            # Overlapping! Remove current interval
            removed += 1
        else:
            # Non-overlapping, update prev_end
            prev_end = end
    
    return removed
```

i = 2 -> height = 2, width = 1, just the first rectangel

        This i is not done. See heights[4] = 2 < heights[2] = 5, so still in while and height = 5, width = 4 - 1 - 1=2.

At last, the 2nd while is for cleaning the stack to make sure there is nothing bigger than the max_area at present.

```python
def largestRectangleArea(heights):
        stack = []
        max_area = 0
        heights.append(0)  # Add sentinel to trigger final cleanup
        
        for i in range(len(heights)):
            while stack and heights[i] < heights[stack[-1]]:
                h_index = stack.pop()
                h = heights[h_index]
                width = i if not stack else i - stack[-1] - 1
                max_area = max(max_area, h * width)
            
            stack.append(i)
        
        return max_area
```

```python
def maximalRectangle(matrix):
    if not matrix or not matrix[0]: ## Empty matrix and matrix with only one 0
        return 0
    
    cols = len(matrix[0])
    heights = [0] * cols
    max_area = 0
    
    for row in matrix:
        # Build histogram heights for this row
        print(f"this is the row here: {row}")
        for j in range(cols):
            if row[j] == '1':
                heights[j] += 1
            else:
                heights[j] = 0 
        
        # Find largest rectangle in this histogram
        max_area = max(max_area, largestRectangleArea(heights))
    
    return max_area


def largestRectangleArea(heights):
    stack = []
    max_area = 0
    heights.append(0)  # Sentinel to process remaining
    
    for i in range(len(heights)):
        while stack and heights[i] < heights[stack[-1]]:
            h_index = stack.pop()
            h = heights[h_index]
            width = i if not stack else i - stack[-1] - 1
            max_area = max(max_area, h * width)
        
        stack.append(i)
    
    heights.pop()  # Remove sentinel
    return max_area


print(maximalRectangle( [['0','1'],['1','1'],['1','1'],['1','0']]))        
```
