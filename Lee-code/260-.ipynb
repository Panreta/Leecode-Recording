# 260. Single Number III

```python
from collections import Counter
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        count = Counter(nums)
        result = []
        elements = set(nums)
        for element in elements:
            if count[element] == 1:
                result.append(element)

        return result
        
```

# 491. Non-decreasing Subsequences

1. choose all the subsequence(if n = len(nums),this is a set of 2^n size)
2. test each of them are sorted or not, only keep the sorted one

```python
def findSubsequences(nums):
    result = set() # don't want duplicated element 
    
    def dfs(start, path):
        if len(path) >= 2:
            result.add(tuple(path))
        
        for i in range(start, len(nums)):
            if not path or nums[i] >= path[-1]:
                dfs(i + 1, path + [nums[i]])
    
    dfs(0, [])
    return list(result)

#[4,6,7,7]
```

```python
def findSubsequences(nums):
    subs = {()}
    for num in nums:
        subs |= {sub + (num,) for sub in subs if not sub or sub[-1] <= num}
    return [sub for sub in subs if len(sub) >= 2]
```

# 264. Ugly Number II

think about a tree with 3 brance is a good way

```python
def nthUglyNumber(n):
    dp = [1] * n
    i2 = i3 = i5 = 0
    
    for i in range(1, n):
        next2 = dp[i2] * 2
        next3 = dp[i3] * 3
        next5 = dp[i5] * 5
        
        dp[i] = min(next2, next3, next5)
        
        if dp[i] == next2: i2 += 1
        if dp[i] == next3: i3 += 1
        if dp[i] == next5: i5 += 1
    
    return dp[-1]
```

# 268. Missing Number

```python
def missingNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        nums = sorted(nums)
        for i, num in enumerate(nums):
            if i != num:
                return i
        return len(nums)
```

# 273. Integer to English Words

```python
def numberToWords(num):
        to19 = 'One Two Three Four Five Six Seven Eight Nine Ten Eleven Twelve ' \
            'Thirteen Fourteen Fifteen Sixteen Seventeen Eighteen Nineteen'.split()
        tens = 'Twenty Thirty Forty Fifty Sixty Seventy Eighty Ninety'.split()
        def words(n):
            if n < 20:
                return to19[n-1:n]
            if n < 100:
                return [tens[n/10-2]] + words(n%10)
            if n < 1000:
                return [to19[n/100-1]] + ['Hundred'] + words(n%100)
            for p, w in enumerate(('Thousand', 'Million', 'Billion'), 1):
                if n < 1000**(p+1):
                    return words(n/1000**p) + [w] + words(n%1000**p)
        return ' '.join(words(num)) or 'Zero'
```

# 274. H-Index

So astonished to find these 2 have the same speed

```python
class Solution(object):
    def hIndex(self, citations):
        """
        :type citations: List[int]
        :rtype: int
        """
        h = 0

        citations.sort(reverse=True)

        for i, c in enumerate(citations):
            if c >= i + 1:
                h = i + 1
            else:
                break

        return h
```

```python
class Solution(object):
    def hIndex(self, citations):
        """
        :type citations: List[int]
        :rtype: int
        """

        citations.sort()
        n = len(citations)
        
        left, right = 0, n
        while left < right:
            mid = (left + right + 1) // 2
            if citations[n - mid] >= mid:
                left = mid # the zone can't be
            else:
                right = mid - 1
        
        return right
```
