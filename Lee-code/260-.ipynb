# 260. Single Number III

```python
from collections import Counter
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        count = Counter(nums)
        result = []
        elements = set(nums)
        for element in elements:
            if count[element] == 1:
                result.append(element)

        return result
        
```

```python
def findSubsequences(nums):
    result = set() # don't want duplicated element 
    
    def dfs(start, path):
        if len(path) >= 2:
            result.add(tuple(path))
        
        for i in range(start, len(nums)):
            if not path or nums[i] >= path[-1]:
                dfs(i + 1, path + [nums[i]])
    
    dfs(0, [])
    return list(result)

#[4,6,7,7]
```

```python
def findSubsequences(nums):
    subs = {()}
    for num in nums:
        subs |= {sub + (num,) for sub in subs if not sub or sub[-1] <= num}
    return [sub for sub in subs if len(sub) >= 2]
```

# 264. Ugly Number II

think about a tree with 3 brance is a good way

```python
def nthUglyNumber(n):
    dp = [1] * n
    i2 = i3 = i5 = 0
    
    for i in range(1, n):
        next2 = dp[i2] * 2
        next3 = dp[i3] * 3
        next5 = dp[i5] * 5
        
        dp[i] = min(next2, next3, next5)
        
        if dp[i] == next2: i2 += 1
        if dp[i] == next3: i3 += 1
        if dp[i] == next5: i5 += 1
    
    return dp[-1]
```

# 268. Missing Number

```python
def missingNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        nums = sorted(nums)
        for i, num in enumerate(nums):
            if i != num:
                return i
        return len(nums)
```

# 274. H-Index

So astonished to find these 2 have the same speed

```python
class Solution(object):
    def hIndex(self, citations):
        """
        :type citations: List[int]
        :rtype: int
        """
        h = 0

        citations.sort(reverse=True)

        for i, c in enumerate(citations):
            if c >= i + 1:
                h = i + 1
            else:
                break

        return h
```

```python
class Solution(object):
    def hIndex(self, citations):
        """
        :type citations: List[int]
        :rtype: int
        """

        citations.sort()
        n = len(citations)
        
        left, right = 0, n
        while left < right:
            mid = (left + right + 1) // 2
            if citations[n - mid] >= mid:
                left = mid # the zone can't be
            else:
                right = mid - 1
        
        return right
```

# 279. Perfect Squares

3 = 1 + 1 + 1
4 = 4

```python
def numSquares(n):
    dp = [n+1 for _ in range(n+1)] # worst case is 1*1 + 1*1 + ... + 1*1 (n times)
    print(dp)
    dp[0] = 0
    dp[1] = 1
    x = 1
    while x*x <= n:
        sq = x*x
        for i in range(sq, n+1): # split out to [x^2,(x+1)^2)]
            dp[i] = min(dp[i-sq]+1, dp[i]) # this is the updating of batch, you can think 5 is just dp[1] + 1(1 + 4)
        x += 1
    #print dp
    return dp[n]

print(numSquares(12))       
```

# 282. Expression Add Operators(hard)

index: Current position in the string

path: The expression string built so far (e.g., "1+2")

value: Current evaluation result (e.g., 3 after 1+2)

prev: The last number added/subtracted (to handle *)

Why prev? When we multiply, we need to "undo" the last operation and redo it with multiplication!

```python
def addOperators(num, target):
        res = []

        def backtrack(index, path, value, prev):
            if index == len(num):
                if value == target:
                    res.append(path)
                return

            for i in range(index, len(num)):
                # Skip numbers with leading zeros
                if i != index and num[index] == '0':
                    break
                curr_str = num[index:i+1]
                print(f"curr_str: {curr_str}")
                curr = int(curr_str)

                if index == 0:
                    backtrack(i+1, curr_str, curr, curr)
                else:
                    # Addition
                    backtrack(i+1, path + '+' + curr_str, value + curr, curr)
                    # Subtraction
                    backtrack(i+1, path + '-' + curr_str, value - curr, -curr)
                    # Multiplication
                    backtrack(i+1, path + '*' + curr_str, value - prev + prev * curr, prev * curr)

        # Ensure num is a string
        num = str(num)
        backtrack(0, "", 0, 0)
        return res

print(addOperators("123", 6))
```

# 283. Move Zeroes

```python
def moveZeroes(nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        lastNonZero = 0
        for i in range(len(nums)):
            if nums[i] != 0:
                nums[lastNonZero],nums[i] = nums[i], nums[lastNonZero]
                lastNonZero += 1

        return nums
```

Maintain the order of 0

```python
def moveZeroes(nums):
    """
    Preserve order of both non-zeros AND zeros.
    Uses O(n) extra space.
    """
    non_zeros = []
    zeros = []
    
    # Separate while preserving order
    for num in nums:
        if num == 0:
            zeros.append(num)
        else:
            non_zeros.append(num)
    
    # Rebuild: non-zeros first, then zeros
    result = non_zeros + zeros
    
    # Copy back to nums (in-place)
    for i in range(len(nums)):
        nums[i] = result[i]
    
    return nums
```
