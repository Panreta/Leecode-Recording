```python
import json

class TrieNode():
    def __init__(self):
        self.children = {}
        self.value = None  # marks end of a word

class Trie():
    def __init__(self):
        self.root = TrieNode()

    def insert(self, key, value):
        node = self.root
        for char in key:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.value = value # if goes to the end, put a value here

    def search(self, word):
        return self.dfs(self.root, word, 0)
    
    def dfs(self, node, word, i):
        # check the length
        if i == len(word):
            return node.value if node.value else "Not Found"
        
        char = word[i]
        if char == '*':
            for child in node.children.values():
                result = self.dfs(child,word,i + 1)
                if result != "Not Found": # not false
                    return result
        else:
            if char not in node.children:
                return "Not Found"
            return self.dfs(node.children[char],word,i+1)
                
        return "Not Found"
        

        
class IdIpProcedure():
    def __init__(self):
        self.table = {}
        

    def stringToJson(self,s): # str -> json
        s = s.split(":")
        key, value = s[0],""
        track = ""
        flag = 0
        for i in range(1,len(s) - 1): # element between
            for char in s[i]:
                if char.isalpha() and flag == 0:# first meet key
                    value = track
                    self.table[key] = value
                    track = ""
                    flag = 1
                track += char
            key = track
            track = ""
            flag = 0
        value = s[-1]
        self.table[key] = value
        return json.dumps(self.table)
    
    def linearSearch(self,id):# input like "a*c*3", output like "value"
        for key in self.table.keys():
            if self.match(key,id):
                return self.table[key]
        return "Not Found"
    
    def match(self,key,id):
        if len(key) != len(id):
            return False
        for i in range(len(key)):
            if id[i] != '*' and key[i] != id[i]:
                return False
        return True
    
    def TrieSearch(self,id):
        trie = Trie()
        for key,value in self.table.items():
            trie.insert(key,value)

        return trie.search(id)
    
if __name__ == "__main__":
    s = "abc123:192.168.1.1xyz789:10.0.0.254user456:172.16.0.1"
    proc = IdIpProcedure()
    print(proc.stringToJson(s))
    print(list(proc.table.keys()))
    print(proc.linearSearch("a*c1*3"))
    print(proc.TrieSearch("a*c1*3"))
```

# Trie Tree

```python
class TrieNode():
    def __init__(self):
        self.children = {}
        self.is_end = False  # marks end of a word

class Trie():
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True

    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return node.is_end
```
