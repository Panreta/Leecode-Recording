```python
def rangeBitwiseAnd(left, right):
        """
        :type left: int
        :type right: int
        :rtype: int
        """
        shift = 0
        # Find common prefix by removing differing bits
        while left < right:
            left >>= 1
            right >>= 1
            shift += 1
        
        # Shift back to restore the common prefix
        return left << shift
```

```python
def isIsomorphic(s, t):
    if len(s) != len(t):
        return False

    mapping_s_to_t = {}
    mapping_t_to_s = {}

    for i in range(len(s)):
        char_s = s[i]
        char_t = t[i]

        if char_s in mapping_s_to_t:
            if mapping_s_to_t[char_s] != char_t:
                return False
        else:
            mapping_s_to_t[char_s] = char_t

        if char_t in mapping_t_to_s:
            if mapping_t_to_s[char_t] != char_s:
                return False
        else:
            mapping_t_to_s[char_t] = char_s

    return True
```

# 207. Should take this course before take another: Kahn algorithm

What does it really want: Can be Topological Sort

```python
from collections import defaultdict, deque

class Solution(object):
    def canFinish(self, numCourses, prerequisites):

        # Build adjacency list and in-degree count
        graph = defaultdict(list)
        in_degree = [0] * numCourses
        
        for course, prereq in prerequisites:
            graph[prereq].append(course)
            in_degree[course] += 1
        
        # Start with courses that have no prerequisites
        queue = deque()
        for i in range(numCourses):
            if in_degree[i] == 0:
                queue.append(i)
        
        # Process courses using topological sort (Kahn's algorithm)
        completed = 0
        
        while queue:
            course = queue.popleft()
            completed += 1
            
            # Reduce in-degree for dependent courses
            for next_course in graph[course]:
                in_degree[next_course] -= 1
                if in_degree[next_course] == 0:
                    queue.append(next_course)
        
        # If we completed all courses, no cycle exists
        return completed == numCourses
```

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None  # Store the complete word at leaf nodes

class Solution(object):
    def findWords(self, board, words):

        # process words into trie(1)
        root = TrieNode()
        for word in words:
            node = root
            for char in word:
                if char not in node.children:
                    node.children[char] = TrieNode()
                node = node.children[char]
            node.word = word

        
        
        m, n = len(board), len(board[0])
        result = []
        
        def dfs(i, j, parent, node):
            char = board[i][j]
            curr_node = node.children[char]
            
            # Found word
            if curr_node.word:
                result.append(curr_node.word)
                curr_node.word = None
            
            # Mark visited
            board[i][j] = '#'
            
            # Explore neighbors
            for di, dj in [(0,1), (1,0), (0,-1), (-1,0)]:
                ni, nj = i + di, j + dj
                if (0 <= ni < m and 0 <= nj < n and 
                    board[ni][nj] in curr_node.children):
                    dfs(ni, nj, curr_node, curr_node)
            
            # Backtrack
            board[i][j] = char
            
            # Optimization: remove leaf node
            if not curr_node.children:
                del node.children[char]
        
        for i in range(m):
            for j in range(n):
                if board[i][j] in root.children:
                    dfs(i, j, root, root)
        
        return result
```

```python
def combinationSum3(k, n):
    result = []
    
    def backtrack(start, path, remaining):
        # Base case 1: Found a valid combination
        if len(path) == k and remaining == 0:
            result.append(path[:])  # Make a copy
            return
        
        # Base case 2: Invalid - stop exploring
        if len(path) == k or remaining < 0:
            return
        
        # Try numbers from start to 9
        for num in range(start, 10):
            # Skip if this number is too large
            if num > remaining:
                break
            
            # Choose
            path.append(num)
            
            # Explore (num+1 to avoid reusing the same number)
            backtrack(num + 1, path, remaining - num)
            
            # Unchoose (backtrack)
            path.pop()# Try [1,2], 2 doesn't match, pop 2, and try [1,3]
    
    backtrack(1, [], n)
    return result

```

# 220. Contains Duplicate III

If the numbers are in a range, that's also count.

```python
def containsNearbyAlmostDuplicate(self, nums, indexDiff, valueDiff):
    """
    
    Strategy: Use bucket sort to group numbers by value ranges.
    If two numbers are close in value (diff <= valueDiff), they'll be
    in the same bucket or adjacent buckets.
    """
    
    
    # Dictionary to store buckets: {bucket_id: number_in_that_bucket}
    # We only store ONE number per bucket (the most recent one in our window)
    bucket_map = {}
    

    bucket_size = valueDiff + 1 ## elements that can be gathered into one bucket
    
    # Iterate through each number with its index
    for current_index, current_number in enumerate(nums):
        
  
        current_bucket_id = current_number // bucket_size
        
        # Check 1: Is there already a number in the SAME bucket?
        # If yes, it means the difference is definitely <= valueDiff
        # (because they're in the same value range)
        if current_bucket_id in bucket_map:
            return True
        
        # Check 2: Is there a number in the LEFT adjacent bucket?
        # Check if the actual value difference is <= valueDiff
        left_bucket_id = current_bucket_id - 1
        if left_bucket_id in bucket_map:
            left_bucket_number = bucket_map[left_bucket_id]
            if abs(current_number - left_bucket_number) <= valueDiff:
                return True
        
        # Check 3: Is there a number in the RIGHT adjacent bucket?
        # Check if the actual value difference is <= valueDiff
        right_bucket_id = current_bucket_id + 1
        if right_bucket_id in bucket_map:
            right_bucket_number = bucket_map[right_bucket_id]
            if abs(current_number - right_bucket_number) <= valueDiff:
                return True
        
        # Add current number to its bucket
        # This overwrites any previous number in this bucket (which is fine,
        # because we only care about the most recent number in our sliding window)
        bucket_map[current_bucket_id] = current_number
        
        # Maintain sliding window: remove the number that's now too far away
        # Only remove if we've processed more than indexDiff elements
        if current_index >= indexDiff:
            # Calculate which bucket the old number was in
            old_number = nums[current_index - indexDiff]
            old_bucket_id = old_number // bucket_size
            
            # Remove the old bucket from our map
            # (This keeps our window size <= indexDiff)
            del bucket_map[old_bucket_id]
    
    # No valid pair found
    return False


```

# 221. Maximal Square(DP)

Why DFS Doesn't Work Well
Problem: Find the largest square (not just any shape) of 1's.
Why DFS fails:

DFS finds connected regions of ANY shape (not just squares)
Checking if a region is a square is complex
Time complexity would be terrible: O(m × n × min(m,n)²)

```python
def maximalSquare(self, matrix):
      

        
        rows = len(matrix)
        cols = len(matrix[0])
        
        # DP table: dp[i][j] = side length of largest square ending at (i,j)
        dp = [[0] * cols for _ in range(rows)]
        
        # Track the maximum side length found
        max_side_length = 0
        
        # Fill the DP table
        for row in range(rows):
            for col in range(cols):
                # If current cell is '1'
                if matrix[row][col] == '1':
                    if row == 0 or col == 0: # At the boundary
                        # First row or first column: can only form 1×1 square
                        dp[row][col] = 1
                    else:
                        # Check top, left, and top-left neighbors
                        # The square size is limited by the smallest neighbor + 1
                        dp[row][col] = min(
                            dp[row - 1][col],      # Above
                            dp[row][col - 1],      # Left
                            dp[row - 1][col - 1]   # Diagonal (top-left)
                        ) + 1
                    
                    # Update maximum side length
                    max_side_length = max(max_side_length, dp[row][col])
        
        # Return area (side length squared)
        return max_side_length * max_side_length
```

The last level is filled from left to right (no gaps allowed)

```python
def computeArea(ax1, ay1, ax2, ay2, bx1, by1, bx2, by2):
    # Calculate area of both rectangles
    areaA = (ax2 - ax1) * (ay2 - ay1)
    areaB = (bx2 - bx1) * (by2 - by1)
    
#   - Start of overlap: max(ax1, bx1)  ← rightmost of the two left edges
#   - End of overlap: min(ax2, bx2)    ← leftmost of the two right edges
# pay attention, there is 8 possible cases, and 3 binary minmax can cover all(2^3 = 8) cases
    overlap_width = max(0, min(ax2, bx2) - max(ax1, bx1))
    overlap_height = max(0, min(ay2, by2) - max(ay1, by1))
    
    # Calculate overlap area
    overlap_area = overlap_width * overlap_height
    
    # Total area covered by both rectangles
    total_area = areaA + areaB - overlap_area
    
    return total_area
    

```

# 229. Majority Element II

```python
from collections import Counter
def majorityElement(nums):
    """
    :type nums: List[int]
    :rtype: List[int]
    """
    n = len(nums)
    threshhold = n // 3
    count = Counter(nums)
    print(f"Count is {count}")

    result = []
    for num, freq in count.items():
        if freq > threshhold:
            result.append(num)

    return result

nums = [3,2,3]
print(majorityElement(nums))
```

# 230. Kth Smallest Element in a BST

```python
def kthSmallest(self, root, k):
    """
    :type root: Optional[TreeNode]
    :type k: int
    :rtype: int
    """

    stack = []
    cur = root
    count = 0

    while cur or stack:
        while cur:
            stack.append(cur)
            cur = cur.left

        # smallest now
        cur = stack.pop()
        count += 1

        if count == k:
            return cur.val

        cur = cur.right
    return  None
```

```python
def kthSmallest(self, root, k):
    """
    Fixed recursive in-order traversal.
    Use a helper to track count across recursive calls.
    """
    def inorder(node):
        if not node:
            return None
        
        # Search left subtree
        left_result = inorder(node.left)
        if left_result is not None:
            return left_result  # Found in left subtree
        
        # Process current node
        self.count += 1
        if self.count == k:
            return node.val  # Found!
        
        # Search right subtree
        return inorder(node.right)
    
    self.count = 0
    return inorder(root)
```

231. Power of Two

```python
n = 16
def isPowerOfTwo(n):
    if n == 1:
        return True
    cur = n
    while cur > 2:
        cur //= 2

    return cur == 2
print(isPowerOfTwo(5))
```

# 240.Search a 2D Matrix II

count from the right-up element.

```python
def searchMatrix(matrix, target):
    m,n = len(matrix), len(matrix[0])
    row, col = 0, n - 1 # for index needed
    while row < m and col >= 0: # for only conduct interio
        cur = matrix[row][col]
        if cur == target:
            return True
        elif cur < target:
            col -= 1
        else:
            row += 1
    return False
```

# 241. Different Ways to Add Parentheses

```python
def diffWaysToCompute(self, expression):
        memo = {}
        
        def dfs(expr):
            if expr.isdigit():
                return [int(expr)]
            
            if expr in memo:
                return memo[expr]
            
            res = []
            
            for i, c in enumerate(expr):
                if c in "+-*":
                    left = dfs(expr[:i])
                    right = dfs(expr[i+1:])
                    
                    for l in left:
                        for r in right:
                            if c == '+':
                                res.append(l + r)
                            elif c == '-':
                                res.append(l - r)
                            else:
                                res.append(l * r)
            
            memo[expr] = res
            return res
        
        return dfs(expression)
```

# 242. Valid Anagram

```python
def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        length_s = len(s)
        length_t = len(t)

        count_s = Counter(s)
        count_t = Counter(t)

        if length_s < length_t:
            for char in t:
                if char not in count_s:
                    return False
                else: # suppose the key is in
                    if count_s[char] != count_t[char]:
                        return False
            return True
        else:
            for char in s:
                if char not in count_t:
                    return False
                else: # suppose the key is in
                    if count_s[char] != count_t[char]:
                        return False
            return True
```

Fastest

```python
def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        if len(s) != len(t):
            return False
        s_set = set(s)
        for i in s_set:
            if s.count(i) != t.count(i):
                return False
        return True
```

# 257. Binary Tree Paths

```python
import time
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def listToBinaryTree(lst):
    if not lst:
        return None
    root = TreeNode(lst[0])
    queue = [root]
    i = 1
    while queue and i < len(lst):
        current = queue.pop(0)
        if lst[i] is not None:
            current.left = TreeNode(lst[i])
            queue.append(current.left)
        i += 1
        if i < len(lst) and lst[i] is not None:
            current.right = TreeNode(lst[i])
            queue.append(current.right)
        i += 1
    return root


```

```python
def binaryTreePaths(root):
    if not root:
        return []
    
    result = []
    stack = [(root, str(root.val))]  # (node, path_so_far)
    
    while stack:
        cur, path = stack.pop()
        
        # Leaf node — save the path
        if not cur.left and not cur.right:
            result.append(path)
        
        if cur.right:
            stack.append((cur.right, path + "->" + str(cur.right.val)))
        if cur.left:
            stack.append((cur.left, path + "->" + str(cur.left.val)))
    
    return result
```

# 258. Add Digits

```python
def sumDigit(n):
    count = 0
    while n:
        residue = n % 10
        count += residue
        n = int((n - residue) // 10)
    return count

print(sumDigit(11))
```
