# 201 

Back to the binary problem now, so we can get a theorem by observation:

**Theorem:** If $left < right$, $\exists k, i$ such that $left_i = 0$, $i < k$, and $\forall i \geq k$, $left_i = right_i$.

```python
def rangeBitwiseAnd(left, right):
        """
        :type left: int
        :type right: int
        :rtype: int
        """
        shift = 0
        # Find common prefix by removing differing bits
        while left < right:
            left >>= 1
            right >>= 1
            shift += 1
        
        # Shift back to restore the common prefix
        return left << shift
```

# Isomorphic: bi-jection, need 2 map to be the same 

```python
def isIsomorphic(s, t):
    if len(s) != len(t):
        return False

    mapping_s_to_t = {}
    mapping_t_to_s = {}

    for i in range(len(s)):
        char_s = s[i]
        char_t = t[i]

        if char_s in mapping_s_to_t:
            if mapping_s_to_t[char_s] != char_t:
                return False
        else:
            mapping_s_to_t[char_s] = char_t

        if char_t in mapping_t_to_s:
            if mapping_t_to_s[char_t] != char_s:
                return False
        else:
            mapping_t_to_s[char_t] = char_s

    return True
```

```python
def reverseList(self, head):
    """
    :type head: Optional[ListNode]
    :rtype: Optional[ListNode]
    """
    prev = None
    cur = head

    while cur:
        post = cur.next
        cur.next = prev

        prev = cur
        cur = post

    return prev
```

# 207. Should take this course before take another: Kahn algorithm

What does it really want: Can be Topological Sort

```python
from collections import defaultdict, deque

class Solution(object):
    def canFinish(self, numCourses, prerequisites):

        # Build adjacency list and in-degree count
        graph = defaultdict(list)
        in_degree = [0] * numCourses
        
        for course, prereq in prerequisites:
            graph[prereq].append(course)
            in_degree[course] += 1
        
        # Start with courses that have no prerequisites
        queue = deque()
        for i in range(numCourses):
            if in_degree[i] == 0:
                queue.append(i)
        
        # Process courses using topological sort (Kahn's algorithm)
        completed = 0
        
        while queue:
            course = queue.popleft()
            completed += 1
            
            # Reduce in-degree for dependent courses
            for next_course in graph[course]:
                in_degree[next_course] -= 1
                if in_degree[next_course] == 0:
                    queue.append(next_course)
        
        # If we completed all courses, no cycle exists
        return completed == numCourses
```

```python
def minSubArrayLen(self, target, nums):
        """
        :type target: int
        :type nums: List[int]
        :rtype: int
        """
        left = 0
        length = float('inf')
        cur_sum = 0

        for right in range(len(nums)):
            cur_sum += nums[right]

            while cur_sum >= target:
                length = min(length,right - left + 1)
                cur_sum -= nums[left]
                left += 1
        if length == float('inf'):
            return 0
        else:
            return length
```

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None  # Store the complete word at leaf nodes

class Solution(object):
    def findWords(self, board, words):

        # process words into trie(1)
        root = TrieNode()
        for word in words:
            node = root
            for char in word:
                if char not in node.children:
                    node.children[char] = TrieNode()
                node = node.children[char]
            node.word = word

        
        
        m, n = len(board), len(board[0])
        result = []
        
        def dfs(i, j, parent, node):
            char = board[i][j]
            curr_node = node.children[char]
            
            # Found word
            if curr_node.word:
                result.append(curr_node.word)
                curr_node.word = None
            
            # Mark visited
            board[i][j] = '#'
            
            # Explore neighbors
            for di, dj in [(0,1), (1,0), (0,-1), (-1,0)]:
                ni, nj = i + di, j + dj
                if (0 <= ni < m and 0 <= nj < n and 
                    board[ni][nj] in curr_node.children):
                    dfs(ni, nj, curr_node, curr_node)
            
            # Backtrack
            board[i][j] = char
            
            # Optimization: remove leaf node
            if not curr_node.children:
                del node.children[char]
        
        for i in range(m):
            for j in range(n):
                if board[i][j] in root.children:
                    dfs(i, j, root, root)
        
        return result
```

```python
def combinationSum3(k, n):
    result = []
    
    def backtrack(start, path, remaining):
        # Base case 1: Found a valid combination
        if len(path) == k and remaining == 0:
            result.append(path[:])  # Make a copy
            return
        
        # Base case 2: Invalid - stop exploring
        if len(path) == k or remaining < 0:
            return
        
        # Try numbers from start to 9
        for num in range(start, 10):
            # Skip if this number is too large
            if num > remaining:
                break
            
            # Choose
            path.append(num)
            
            # Explore (num+1 to avoid reusing the same number)
            backtrack(num + 1, path, remaining - num)
            
            # Unchoose (backtrack)
            path.pop()# Try [1,2], 2 doesn't match, pop 2, and try [1,3]
    
    backtrack(1, [], n)
    return result

```

```python
import heapq

heap = []
heapq.heappush(heap, 5)
heapq.heappush(heap, 3)
heapq.heappush(heap, 7)
heapq.heappush(heap, 1)

print(heap)  # [1, 3, 7, 5] - min heap property maintained

## Pop: get the smallest element, which is defaultly the leftmost element
heap = [1, 3, 7, 5]

smallest = heapq.heappop(heap)
print(smallest)  # 1
print(heap)      # [3, 5, 7]

## Pushpop: pop out the smallest element, then push the new item, and return the popped smallest element
heap = [1, 3, 7, 5]

# More efficient than push then pop separately
result = heapq.heappushpop(heap, 2)
print(result)  # 1 (smallest of old heap and new item)
print(heap)    # [2, 3, 7, 5]
```

```python
import heapq
from collections import Counter

class Solution(object):
    def getSkyline(self, buildings):
        events = []
        
        for left, right, height in buildings:
            # Start: (x, 0, -height) - negative for tallest first
            events.append((left, 0, -height))
            # End: (x, 1, height) - positive for shortest first
            events.append((right, 1, height))
        
        # Sort by: x coordinate, then event type, then height
        events.sort()
        
        result = []
        max_heap = [0]
        active = Counter([0])
        
        i = 0
        while i < len(events):
            curr_x = events[i][0]## all the left
            
            # Process ALL events at the same x-coordinate together
            while i < len(events) and events[i][0] == curr_x:
                x, typ, h = events[i]
                
                if typ == 0:  # Start (h is negative)
                    heapq.heappush(max_heap, h)
                    active[h] += 1
                else:  # End (h is positive, need to negate)
                    active[-h] -= 1
                    if active[-h] == 0:
                        del active[-h]
                
                i += 1
            
            # Clean up stale heights
            while max_heap[0] not in active:
                heapq.heappop(max_heap)
            
            curr_max = -max_heap[0]
            
            # Only add if height changed
            if not result or result[-1][1] != curr_max:
                result.append([curr_x, curr_max])
        
        return result


## Key Changes:

```
