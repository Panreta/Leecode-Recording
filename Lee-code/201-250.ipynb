# 201 

Back to the binary problem now, so we can get a theorem by observation:

**Theorem:** If $left < right$, $\exists k, i$ such that $left_i = 0$, $i < k$, and $\forall i \geq k$, $left_i = right_i$.

```python
def rangeBitwiseAnd(left, right):
        """
        :type left: int
        :type right: int
        :rtype: int
        """
        shift = 0
        # Find common prefix by removing differing bits
        while left < right:
            left >>= 1
            right >>= 1
            shift += 1
        
        # Shift back to restore the common prefix
        return left << shift
```

```python
def isHappy(self, n):
    """
    :type n: int
    :rtype: bool
    """
    seen = set()
    while True:
        if n in seen:
            return False
        seen.add(n)
        n = str(n)
        res = 0
        for char in n:
            res += int(char)**2
        if res == 1:
            return True
        n = res
```

# 203 Linked List

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def listToLinkedList(lst):
    if not lst:
        return None
    head = ListNode(lst[0])
    current = head
    for value in lst[1:]:
        current.next = ListNode(value)
        current = current.next
    return head

def printLinkedList(head):
    current = head
    while current:
        print(current.val, end=' ')
        current = current.next
    print()
```

```python
def removeElements(head, val):
    ## CC
    if not head:
        return head

    # cut head
    while head and head.val == val:
        head = head.next

    if not head:
        return head

    # 1st node is not val
    cur = head
    fast = head.next
    while cur and fast:
        print(f"cur is {cur.val}, fast is {fast.val}")
        while fast and fast.val == val:                
            # fast is in the middle
            cur.next = fast.next
            fast = fast.next

        if fast:
            cur = cur.next
            fast = fast.next
            # fast is the end

    return head
```

```python
list = [1,2,6,3,4,5,6]
head = listToLinkedList(list)
new_head = removeElements(head = head,val = 6)
printLinkedList(new_head)
```

```python
def countPrimes(n):
        if n <= 2:
            return 0
        
        # Use list instead of array for faster operations
        isPrime = [True] * n
        isPrime[0] = isPrime[1] = False
        
        # Mark even numbers
        isPrime[4:n:2] = [False] * len(isPrime[4:n:2])
        
        # Process odd numbers only
        for i in range(3, int(n**0.5) + 1, 2):
            if isPrime[i]:
                # Use slice assignment (faster than loop)
                isPrime[i*i:n:2*i] = [False] * len(isPrime[i*i:n:2*i])
        
        return sum(isPrime)
```

# Isomorphic: bi-jection, need 2 map to be the same 

```python
def isIsomorphic(s, t):
    if len(s) != len(t):
        return False

    mapping_s_to_t = {}
    mapping_t_to_s = {}

    for i in range(len(s)):
        char_s = s[i]
        char_t = t[i]

        if char_s in mapping_s_to_t:
            if mapping_s_to_t[char_s] != char_t:
                return False
        else:
            mapping_s_to_t[char_s] = char_t

        if char_t in mapping_t_to_s:
            if mapping_t_to_s[char_t] != char_s:
                return False
        else:
            mapping_t_to_s[char_t] = char_s

    return True
```
