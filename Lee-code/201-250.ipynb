# 201 

Back to the binary problem now, so we can get a theorem by observation:

**Theorem:** If $left < right$, $\exists k, i$ such that $left_i = 0$, $i < k$, and $\forall i \geq k$, $left_i = right_i$.

```python
def rangeBitwiseAnd(left, right):
        """
        :type left: int
        :type right: int
        :rtype: int
        """
        shift = 0
        # Find common prefix by removing differing bits
        while left < right:
            left >>= 1
            right >>= 1
            shift += 1
        
        # Shift back to restore the common prefix
        return left << shift
```

# 202. Happy Number

First I think about changing it to be a vector then find the 2-norm of this. But actually, string is more powerful then vector in algorithm, so changing to string is way better.

```python
def isHappy(self, n):
    """
    :type n: int
    :rtype: bool
    """
    seen = set()
    while True:
        if n in seen:
            return False
        seen.add(n)
        n = str(n)
        res = 0
        for char in n:
            res += int(char)**2
        if res == 1:
            return True
        n = res
```
