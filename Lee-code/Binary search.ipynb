```python
def searchInsert(nums, target):
    left, right = 0 , len(nums) - 1
    while left <= right:
        mid = (left + right)//2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
            print(f"left:{left}")
        else:
            right = mid - 1
            print(f"right{right}")

    return left
nums = [1,3,5,6]
target = 7

print(searchInsert(nums,target))
```
