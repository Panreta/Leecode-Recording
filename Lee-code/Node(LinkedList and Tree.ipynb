## 141 list cycle(fast and slow pointer actually)

```python
def hasCycle(head):
    """
    :type head: ListNode
    :rtype: bool
    """
    if not head or not head.next:# no or just one
        return False

    slow = head
    fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:
            return True

    return False


```

## 142: Two pointer and also some math

it need to find the equivolent yourself, you can see that by this algorithm the pointers will finally meet at the beginning

Let's say the distance from head to cycle start is k
When slow and fast meet inside the cycle, slow has traveled k + m steps
Fast has traveled k + m + nC (where C is cycle length, n is number of loops)
Since fast is 2x speed: 2(k + m) = k + m + nC
Simplifying: k = nC - m

Consider reset slow now, the distance slow need to go is nC-m, and fast is C-m. Also, if slow has passed C distance, which left 
(n-1)C-m, for fast since it has the same speed, it has still C-m to go since it is in the circle.

```python
def detectCycle(head):
    if not head or not head.next:
        return None
    
    # Phase 1: Detect if cycle exists
    slow = head
    fast = head
    
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        
        if slow == fast:
            # Cycle detected! Now find the start
            # Phase 2: Find cycle start
            slow = head  # Reset slow to head
            while slow != fast: # fast is in the circle now
                slow = slow.next
                fast = fast.next  # Move both at same speed
            
            return slow  # They meet at cycle start
    
    return None  # No cycle
```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def reverseBetween(self, head, left, right):
        """
        :type head: Optional[ListNode]
        :type left: int
        :type right: int
        :rtype: Optional[ListNode]
        """

        if not head or left == right:
            return head

        dummy = ListNode(0)
        dummy.next = head

        # finding the start
        prev = dummy
        for _ in range(left - 1):
            prev = prev.next

        # building a list
        stack = []
        cur = prev.next
        for _ in range(right - left + 1):
            stack.append(cur)
            cur = cur.next


        # build back the linkList
        while stack:
            prev.next = stack.pop()
            prev  = prev.next
        prev.next = cur

        return dummy.next
        
```

```python
def isValidBST(root):
    def helper(node, min_val, max_val):
        if not node:
            return True
        
        if node.val <= min_val or node.val >= max_val:
            return False
        
        # Left subtree: all values < node.val
        # Right subtree: all values > node.val
        return (helper(node.left, min_val, node.val) and
                helper(node.right, node.val, max_val))
    
    return helper(root, float('-inf'), float('inf'))
```

```python
def recoverTree(self, root):
        self.first = None
        self.second = None
        self.prev = None
        
        def inorder(node):
            if not node:
                return
            
            inorder(node.left)
            
            # Detect swapped nodes
            if self.prev and self.prev.val > node.val:
                if not self.first:
                    self.first = self.prev
                self.second = node
            self.prev = node

            inorder(node.right)
        
        inorder(root)
        
        # Swap the values of the two incorrect nodes
        self.first.val, self.second.val = self.second.val, self.first.val
```

# 100. Same tree

```python
def isSameTree(self, p, q):
        """
        :type p: Optional[TreeNode]
        :type q: Optional[TreeNode]
        :rtype: bool
        """

        #CC
        # Both empty
        if not p and not q:
            return True
        
        # One empty, one not
        if not p or not q:# symmetric difference here, since I have checked the intersection above
            return False

        if p.val != q.val:
            return False

        return self.isSameTree(p.left,q.left) and self.isSameTree(p.right,q.right)
```

```python
from collections import deque

def levelOrder(root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[List[int]]
        """
        if not root: return []
        q = deque()
        q.append(root)
        f = []
        while q:
            level = len(q)
            a = []
            while level >0:
                z = q.popleft()
                if z.left : q.append(z.left)
                if z.right: q.append(z.right)
                level -= 1
                a.append(z.val)
            f.append(a)
        return f
```

# 108. If balance, first know how to find the height, which is a maximum structure for a node.

```python
class Solution:
    def isBalanced(self, root):
        if not root:
            return True

        def height(node):
            if not node:
                return 0
            return 1 + max(height(node.left), height(node.right))

        r = height(root.right)
        l = height(root.left)
        if abs(r - l) > 1:
            return False
        return self.isBalanced(root.left) and self.isBalanced(root.right)
```

You must really clever on how to use (pre,in,post)-order, 

in-order: print a BST acsendingly, like 1,2,3,...

necessary condition: root comes in the middle

pre-order: print as a root-walking way. 

Post-order: print each level from the leaf to the root, root comes at very last.

```python
def flatten(self, root):
        """
        :type root: TreeNode
        :rtype: None
        """
        curr = root

        while curr:
            if curr.left:
                # Find rightmost node of left subtree
                prev = curr.left
                while prev.right:
                    prev = prev.right

                # Rewire pointers
                prev.right = curr.right
                curr.right = curr.left
                curr.left = None

            curr = curr.right
```

```python
def preorderTraversal(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[int]
        """
        result = []
        if root:
            result.append(root.val)  # append, not extend (root.val is an integer)
            result.extend(self.preorderTraversal(root.left))  # collect left subtree
            result.extend(self.preorderTraversal(root.right))  # collect right subtree

        return result
```
