```python
def romanToInt(s):
    """
    Convert Roman numeral string to integer
    
    Examples:
    - "III" -> 3
    - "IV" -> 4
    - "IX" -> 9
    - "LVIII" -> 58
    - "MCMXCIV" -> 1994
    """
    # Mapping of Roman numerals to integers
    roman_map = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
    }
    
    total = 0
    n = len(s)
    
    for i in range(n):
        # If current value is less than next value, subtract it
        # (handles cases like IV, IX, XL, XC, CD, CM)
        if i < n - 1 and roman_map[s[i]] < roman_map[s[i + 1]]:
            print("it is used")
            total -= roman_map[s[i]]
        else:
            # Otherwise, add it
            total += roman_map[s[i]]
    
    return total
print()

# Test cases
# print(romanToInt("III"))        # 3
print(romanToInt("IV"))         # 4 
# print(romanToInt("IX"))         # 9
# print(romanToInt("LVIII"))      # 58 (L=50, V=5, III=3)
# print(romanToInt("MCMXCIV"))    # 1994 (M=1000, CM=900, XC=90, IV=4)
```

```python
def longestCommonPrefix(strs):
    """
    Zip all strings together and check if all characters at each position match
    """
    if not strs:
        return ""
    
    result = ""
    
    # zip(*strs) creates tuples of characters at each position
    for chars in zip(*strs):
        print(chars)
        # If all characters are the same, add to result
        if len(set(chars)) == 1:
            result += chars[0]
        else:
            break
    
    return result


# Test
print(longestCommonPrefix(["flfwer","flfw","flfght"]))  # "fl"
```

```python
def isValid(s):
    stack = []

    for char in s:
        # treat if this is the bracket you want
        if char == '(' or char == '{' or char == '[':
            stack.append(char)

        else: # like @#$% or something like this or )}] or empty
            if not stack: 
                return False
            
            top = stack.pop()

            # check if it can form a couple
            if char == ')' and top != '(':
                return False
            if char == '}' and top != '{':
                return False
            if char == ']' and top != '[':
                return False
            
    return len(stack) == 0
            
```

```python
def generateParenthesis(n):
    result = []
    
    def backtrack(current_string, open_count, close_count):

        if len(current_string) == 2 * n:
            result.append(current_string)
            return
        
        if open_count < n:
            backtrack(current_string + "(", open_count + 1, close_count)
        
        if close_count < open_count:
            backtrack(current_string + ")", open_count, close_count + 1)
    
    backtrack("", 0, 0)
    return result


# Test
print(generateParenthesis(1))  # ["()"]
print(generateParenthesis(2))  # ["(())", "()()"]
print(generateParenthesis(3))  # 5 combinations
```

```python
def strStr(haystack, needle):
    """
    Manually search for needle in haystack
    """
    
    n = len(haystack)
    m = len(needle)
    
    # If needle is longer than haystack, impossible
    if m > n:
        return -1
    
    # Check each possible starting position
    for i in range(n - m + 1):
        # Check if substring matches
        if haystack[i:i + m] == needle:
            return i
    
    return -1


# Test
print(strStr("sadbutsad", "sad"))    # 0
print(strStr("leetcode", "leeto"))   # -1
print(strStr("hello", "ll"))         # 2
```

```python
def isMatch(s, p):
    """
    Regular expression matching with '.' and '*'
    
    '.' matches any single character
    '*' matches zero or more of the preceding element
    """
    m, n = len(s), len(p)
    
    # dp[i][j] = True if s[0:i] matches p[0:j]

    dp = [[False] * (n + 1) for _ in range(m + 1)]
    print(f"original dp is {dp}")
    
    # Empty string matches empty pattern
    dp[0][0] = True
    
    # Handle patterns like a*, a*b*, a*b*c* that can match empty string
    for j in range(2, n + 1):
        if p[j - 1] == '*':
            dp[0][j] = dp[0][j - 2]
    
    # Fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            # Current characters in s and p
            s_char = s[i - 1]
            p_char = p[j - 1]
            
            if p_char == '*':
                # '*' can match zero or more of preceding element
                prev_p_char = p[j - 2]
                
                # Case 1: Use '*' to match zero occurrences (ignore prev_char*)
                dp[i][j] = dp[i][j - 2]
                
                # Case 2: Use '*' to match one or more occurrences
                # Only if prev_char matches current s_char
                if prev_p_char == '.' or prev_p_char == s_char:
                    dp[i][j] = dp[i][j] or dp[i - 1][j]
            
            elif p_char == '.' or p_char == s_char:
                # '.' matches any character, or characters match exactly
                dp[i][j] = dp[i - 1][j - 1]
    
    return dp[m][n]


# Test cases
print(isMatch("aa", "a"))         # False
print(isMatch("aa", "a*"))        # True
print(isMatch("ab", ".*"))        # True
print(isMatch("aab", "c*a*b"))    # True
print(isMatch("mississippi", "mis*is*p*."))  # False
```
