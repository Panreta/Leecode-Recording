# 3. longest substring

```python
def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        if len(s) == 0:
            return 0

        # s is 1 to 5e4
        max = 0
        count = 1
        origin = s[0]

        for i in range(1,len(s)):
            if s[i] not in origin:
                count += 1
                origin += s[i]
            else:
                if count > max:
                    max = count
                idx = origin.index(s[i])
                origin = origin[1 + idx :] + s[i]
                count = len(origin)
        if count > max:
            max = count
            
        return max
```

# 67. Add Binary

```python
class Solution(object):
    def addBinary(self, a, b):
        """
        :type a: str
        :type b: str
        :rtype: str
        """
        carry = 0
        result = ''
        i = len(a) - 1
        j = len(b) - 1

        while i >= 0 or j >= 0 or carry:

            if i >= 0:
                digita = int(a[i])
            else:
                digita = 0

            if j >= 0:
                digitb = int(b[j])
            else:
                digitb = 0
            cur = digita + digitb + carry

            result += str(cur % 2)
            carry = cur // 2
            i -= 1
            j -= 1
        return result[::-1] 
```

```python
def twoSum(nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        # Hash table to store {value: index}
        seen = {}
        
        for i in range(len(nums)):
            complement = target - nums[i]
            
            # Check if complement exists in hash table
            if complement in seen:
                return [seen[complement], i]
            
            # Store current number and its index
            seen[nums[i]] = i
        
        return []  # No solution found
```

# 299. Bulls and Cows

```python
def getHint(secret, guess):
        """
        :type secret: str
        :type guess: str
        :rtype: str
        """
        bulls = 0
        cows = 0
        
        # Hash tables to count digits (excluding bulls)
        secret_count = {}
        guess_count = {}
        
        # First pass: count bulls and collect non-bull digits
        for i in range(len(secret)):
            if secret[i] == guess[i]:
                bulls += 1
            else:
                # Store non-bull digits for cow counting
                secret_count[secret[i]] = secret_count.get(secret[i], 0) + 1
                guess_count[guess[i]] = guess_count.get(guess[i], 0) + 1
        
        # Second pass: count cows from non-bull digits
        for digit in guess_count:
            if digit in secret_count:
                # Cows = minimum of counts in both
                cows += min(secret_count[digit], guess_count[digit])
        
        return str(bulls) + "A" + str(cows) + "B"
```

# 209

```python
class Solution(object):
    def minSubArrayLen(self, target, nums):
        """
        :type target: int
        :type nums: List[int]
        :rtype: int
        """
        left = 0
        current_sum = 0
        min_length = float('inf')  # Initialize to infinity
        
        for right in range(len(nums)):
            current_sum += nums[right]
            
            # Shrink window: while sum >= target
            while current_sum >= target:# [1,1,1,10000] you don't only get out the first 1 but the rest of them, keep doing, so while
                # Update minimum length
                min_length = min(min_length, right - left + 1)
                
                # Remove left element and move left pointer
                current_sum -= nums[left]
                left += 1
        
        # Return 0 if no valid subarray found
        return min_length if min_length != float('inf') else 0
```

# 904. Fruit Into Baskets

```python
def totalFruit(fruits):
    
    basket = {}  # {fruit_type: count}
    left = 0
    max_fruits = 0
    
    for right in range(len(fruits)):
        # Add current fruit to basket
        fruit = fruits[right]
        basket[fruit] = basket.get(fruit, 0) + 1
        
        while len(basket) > 2:
            # Remove leftmost fruit
            left_fruit = fruits[left]
            basket[left_fruit] -= 1
            
            # If count becomes 0, remove from basket
            if basket[left_fruit] == 0:
                del basket[left_fruit]
            
            left += 1
        
        # Update maximum
        max_fruits = max(max_fruits, right - left + 1)
    
    return max_fruits
```

```python
def subarraySum(nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """
        count = 0
        prefix_sum = 0
        sum_freq = {0: 1}  # Store: {prefix_sum: frequency}
        
        for num in nums:
            prefix_sum += num

            
            # Check if we can form a subarray with sum = k
            if prefix_sum - k in sum_freq:
                count += sum_freq[prefix_sum - k]
            
            # Store current prefix_sum in hash map
            if prefix_sum in sum_freq:
                sum_freq[prefix_sum] += 1
                print(f"here is the sum_freq{sum_freq}")
            else:
                sum_freq[prefix_sum] = 1
        
        return count

print(subarraySum(nums = [1,1,1], k = 2))
```

```python
def findMaxLength(nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        # Hash table: {cumulative_sum: first_index_where_this_sum_occurred}
        sum_index = {0: -1}  
        max_length = 0
        cumulative_sum = 0
        
        for i in range(len(nums)):
            # Treat 0 as -1, and 1 as +1
            if nums[i] == 0:
                cumulative_sum -= 1
            else:
                cumulative_sum += 1
            
            # If we've seen this sum before
            if cumulative_sum in sum_index:
                # Length = current index - previous index
                length = i - sum_index[cumulative_sum]
                max_length = max(max_length, length)
            else:
                # First time seeing this sum, store the index
                sum_index[cumulative_sum] = i
        
        return max_length

print(findMaxLength([0,0,1,1,1,0]))
```

# 56. Merge Intervals

```python
def merge(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: List[List[int]]
        """
        intervals.sort(key = lambda x: x[0])
        merged = []

        for interval in intervals:
            starti, endi = interval

            if not merged or starti > merged[-1][1]:#seperate interval
                merged.append(interval)
            else:
                merged[-1][1] = max(merged[-1][1], endi)
        return merged
```
