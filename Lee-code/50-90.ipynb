```python
def maxSubArray(nums):
    
    max_sum = nums[0]
    current_sum = 0
    
    for num in nums:
        current_sum += num
        max_sum = max(max_sum, current_sum)
        
        # If current_sum becomes negative, reset it
        if current_sum < 0:
            current_sum = 0
    
    return max_sum
```

```python
def spiralOrder(matrix):

    
    result = []
    top, bottom = 0, len(matrix) - 1
    left, right = 0, len(matrix[0]) - 1
    
    while top <= bottom and left <= right:
        # read from left to right
        for col in range(left, right + 1):
            result.append(matrix[top][col])
        top += 1
        
        # read from top to bottom
        for row in range(top, bottom + 1):
            result.append(matrix[row][right])
        right -= 1
        
        # Traverse left along bottom row (if there's a row remaining)
        if top <= bottom:
            for col in range(right, left - 1, -1):
                result.append(matrix[bottom][col])
            bottom -= 1
        
        # Traverse up along left column (if there's a column remaining)
        if left <= right:
            for row in range(bottom, top - 1, -1):
                result.append(matrix[row][left])
            left += 1
    
    return result
```

```python
def canJumpSimple(nums):
    """
    Simpler greedy approach: track the farthest reachable position.
    
    Time Complexity: O(n)
    Space Complexity: O(1)
    """
    max_reach = 0
    
    for i in range(len(nums)):
        # If current position is beyond max reach, we can't get here

        if i > max_reach:
            
            return False
        

        
        # Update the farthest position we can reach
        max_reach = max(max_reach, i + nums[i])
        
        # If we can reach or pass the last index
        if max_reach >= len(nums) - 1:
            return True
    
    return True

nums = [2,3,1,1,4]
print(canJumpSimple(nums))

```

```python
def merge(intervals):
    if not intervals:
        return []
    
    # Sort intervals by start time
    intervals.sort(key=lambda x: x[0])# sorting by the first element of each list
    print(f"Here is the intervals:{intervals}")
    
    merged = []
    
    for interval in intervals:
        start_i, end_i = interval
        
        # If merged is empty or no overlap with last interval
        # No overlap: start_i > end_j (previous end)
     
        if not merged or start_i > merged[-1][1]:#last list, second element
            merged.append(interval)
        else:
            # Overlap: start_i <= end_j
            # Merge by extending the end
            merged[-1][1] = max(merged[-1][1], end_i)
    
    return merged
```

```python


def uniquePaths(m, n):
    # C(m+n-2, m-1) = (m+n-2)! / ((m-1)! * (n-1)!)
    # Optimize: C(m+n-2, min(m-1, n-1))
    
    total = m + n - 2
    k = min(m - 1, n - 1)
    
    result = 1
    for i in range(k):
        result = result * (total - i) // (i + 1)
    
    return result

print(uniquePaths(3,7))
```

```python
def minPathSum(grid):
    m,n = len(grid),len(grid[0])
    dp = [[0] * n for i in range(m)]
    dp[0][0] = grid[0][0]

    for j in range(1,n):
        dp[0][j] = dp[0][j - 1] + grid[0][j]

    for i in range(1,m):
        dp[i][0] = dp[i - 1][0] + grid[i][0]

    for i in range(1,m):
        for j in range(1,n):
            dp[i][j] = grid[i][j] + min(dp[i - 1][j], dp[i][j - 1])
    return dp[m-1][n-1]
```

```python
def isScramble(s1, s2):
    # Memoization cache
    memo = {}
    
    def helper(s1, s2):
        # Base cases
        if s1 == s2:
            return True
        
        if (s1, s2) in memo:
            return memo[(s1, s2)]
        
        # Quick check: same characters?
        if sorted(s1) != sorted(s2):
            memo[(s1, s2)] = False
            return False
        
        n = len(s1)
        
        # Try all possible split points
        for i in range(1, n):
            # Case 1: No swap
            if helper(s1[:i], s2[:i]) and helper(s1[i:], s2[i:]):
                memo[(s1, s2)] = True
                return True
            
            # Case 2: Swap
            if helper(s1[:i], s2[n-i:]) and helper(s1[i:], s2[:n-i]):
                memo[(s1, s2)] = True
                return True
        
        memo[(s1, s2)] = False
        return False
    
    return helper(s1, s2)
```

```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def generateTrees( n):
    # If n is 0, the problem usually expects an empty list
    if n == 0:
        return []
    
    # Memoization dictionary to store results for (start, end) pairs
    memo = {}

    def build(start, end):
        # Base case: if start > end, there's no node to form a tree.
        # We return [None] so the nested loops can still execute once.
        if start > end:
            return [None]
        
        if (start, end) in memo:
            return memo[(start, end)]
        
        all_trees = []
        
        # Try every number in the range as the root
        for i in range(start, end + 1):
            # Generate all possible left and right subtrees
            left_trees = build(start, i - 1)
            right_trees = build(i + 1, end)
            
            # Connect the root i with every combination of left/right subtrees
            for left in left_trees:
                for right in right_trees:
                    root = TreeNode(i)
                    root.left = left
                    root.right = right
                    all_trees.append(root)
        
        memo[(start, end)] = all_trees
        return all_trees

    return build(1, n)
```
