# 20. Valid Parentheses

() — parentheses (or "round brackets")
[] — brackets (or "square brackets")
{} — braces (or "curly brackets")

Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

An input string is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.
Every close bracket has a corresponding open bracket of the same type.

)[(]{}

```python
def isValid(s):
    stack = []
    mapping = {')':'(','}':'{',']':'['} # ()
    

    for char in s: # s = easiest:() hardeast:([]) nested, wrong case:(]
        if char in mapping:
            if not stack or stack.pop() != mapping[char]: # not stack: ) but nothing in stack to match
                return False
        else:
            stack.append(stack)

    return not stack
```

# 199. Binary Tree Right Side View

Given the root of a binary tree, imagine yourself standing on the right side of it, return the values of the nodes you can see ordered from top to bottom.

```python
from collections import deque
def rightSideView(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        ## CC
        if not root:
            return [] 

        queue = deque([root])
        result = []

        while queue:

            level_size = len(queue) #2

            for i in range(level_size):
                node = queue.popleft() #[3,10]

                # the last is the rightmost
                if i == level_size - 1:
                    result.append(node.val)

                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)

        return result

# summary
```

```python
class MinStack(object):
    
    def __init__(self):
        self.stack = []
        self.min_stack = []
        

    def push(self, val): # only choose the smallest one to be on the top of stack
    
        self.stack.append(val)
        if not self.min_stack or val <= self.min_stack[-1]: 
            self.min_stack.append(val)
        

    def pop(self):
  
        if self.stack:
            popped = self.stack.pop()
            if popped == self.min_stack[-1]:
                self.min_stack.pop() # only when the element is the same as the min one then we pop

    def top(self):
        """
        :rtype: int
        """
        return self.stack[-1] if self.stack else None

        

    def getMin(self):
        """
        :rtype: int
        """
        return self.min_stack[-1] if self.min_stack else None

        


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

# 793. Daily Tempreture

Given an array of integers temperatures represents the daily temperatures, return an array answer such that answer[i] is the number of days you have to wait after the ith day to get a warmer temperature. If there is no future day for which this is possible, keep answer[i] == 0 instead.

This is a monotone stack problem, if you see the stack, it always maintains the sequence with monotone property

```python
def dailyTemperatures(temperatures):
    n = len(temperatures)
    result = [0] * n
    stack = []
    

    for i in range(n):
        flag = 1

        while stack:
            print(f"stack before pop: {stack},current temp: {temperatures[i]}, flag: {flag},result: {result}")
            item = stack.pop()
            if item < temperatures[i]:
                result[temperatures.index(item)] += flag
                # stack.append(temperatures[i])
                flag += 1 
            elif item == temperatures[i]:
                break

            else: # less than
                stack.append(item)
                stack.append(temperatures[i])
                result[temperatures.index(item)] += 1

        stack.append(temperatures[i])

    return result

temperatures = [73, 74, 75, 71, 69, 72, 76, 73]
print(dailyTemperatures(temperatures))
```

```python
def dailyTemperatures(temperatures):
        """
        :type temperatures: List[int]
        :rtype: List[int]
        """
        n = len(temperatures)
        result = [0] * n
        stack = []

        for i in range(n):
            while stack and temperatures[i] > temperatures[stack[-1]]:
                pre_idx = stack.pop()
                result[pre_idx] = i - pre_idx

            stack.append(i)

        return result
```

```python
from collections import deque

def maxSlidingWindow(nums, k):
    result = []
    deq = deque()
    
    for i in range(len(nums)):
        
        while deq and deq[0] < i - k + 1:
            deq.popleft()
        
        
        while deq and nums[deq[-1]] < nums[i]:
            deq.pop()
        
        deq.append(i)
        
        
        if i >= k - 1:
            result.append(nums[deq[0]])
    
    return result
```

# 102. Binary Tree Level Order Traversal

```python
def levelOrder(root):

        if not root:
            return []

        q = deque()
        q.append(root)
        final = []
        while q:
            level = len(q)
            apt = []

            while level > 0:
                z = q.popleft()
                if z.left: q.append(z.left)
                if z.right: q.append(z.right)

                level -= 1
                apt.append(z.val)

            final.append(apt)
        return final
```

```python
from collections import deque, defaultdict


def ladderLength(beginWord, endWord, wordList):
    """
    :type beginWord: str
    :type endWord: str
    :type wordList: List[str]
    :rtype: int
    """
    if endWord not in wordList:
        return 0
    
    # Add beginWord to wordList for processing
    wordList.append(beginWord)
    
    # Build adjacency list using pattern matching
    neighbors = defaultdict(list)
    for word in wordList:
        for i in range(len(word)):
            pattern = word[:i] + "*" + word[i+1:] # read letter by letter and replace with *
            neighbors[pattern].append(word)
    
    print(f"neighbors: {dict(neighbors)}")
    # BFS
    queue = deque([beginWord])
    visited = {beginWord}
    length = 1

    while queue:
        for _ in range(len(queue)):
            word = queue.popleft()
            
            if word == endWord: # end condition
                return length
            
            # Try all possible transformations
            for i in range(len(word)):
                pattern = word[:i] + "*" + word[i+1:]
                for neighbor in neighbors[pattern]:
                    if neighbor not in visited:
                        visited.add(neighbor)
                        queue.append(neighbor)
        
        length += 1
    
    return 0
beginWord = "hit"
endWord = "cog"
wordList = ["hot","dot","dog","lot","log","cog"]
print(ladderLength(beginWord, endWord, wordList))
```
