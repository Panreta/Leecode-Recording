```python
def countRangeSum(self, nums, lower, upper):
        prefix = [0] * (len(nums) + 1)
        for i in range(len(nums)):
            prefix[i + 1] = prefix[i] + nums[i] # summation of the prvious elements

        count = [0]

        def merge_sort(arr):
            if len(arr) <= 1:
                return arr

            mid = len(arr) // 2
            left  = merge_sort(arr[:mid])
            right = merge_sort(arr[mid:])

            j = k = 0
            for val in left:
                while j < len(right) and right[j] - val < lower:
                    j += 1
                while k < len(right) and right[k] - val <= upper:
                    k += 1
                count[0] += k - j

            return sorted(left + right)
```

```python
import random
from bisect import bisect_left

class Solution(object):

    def __init__(self, w):
        """
        :type w: List[int]
        """
        self.prefix_sums = []
        running_sum = 0
        for weight in w: #[1,3]
            running_sum += weight
            self.prefix_sums.append(running_sum)
        self.total = running_sum

    def pickIndex(self):
        """
        :rtype: int
        """
        target = random.randint(1,self.total)#[1,self.total]
        return bisect_left(self.prefix_sums,target)
        
```

If w = [1,3,7,9] -> prefix_sums = [1,4,11,20],total = 20; The int'll be picked between [1,20], for [2,4], it will be #{2,3,4}/20 = 3/20 = 3/(1+3+7+9)
