```python
def nextPermutation(nums):
    # Step 1: Find the first decreasing element from the right
    i = len(nums) - 2 
    while i >= 0 and nums[i] >= nums[i + 1]:
        i -= 1
    
    if i >= 0:  # If not the largest permutation
        # Step 2: Find the smallest element larger than nums[i]
        j = len(nums) - 1
        while nums[j] <= nums[i]:
            j -= 1
        
        # Step 3: Swap them
        nums[i], nums[j] = nums[j], nums[i]
    
    # Step 4: Reverse everything after position i
    nums[i + 1:] = reversed(nums[i + 1:])
    print(nums)

nums = [1,2,3,4,5]

for i in range(120):
    nextPermutation(nums)
```

```python
def twoSum(nums, target):
    d = {}
    for i, x in enumerate(nums):
        if target - x in d:
            return [d[target - x], i]
        d[x] = i
nums = [2,7,11,15]
target = 9
print(twoSum(nums,target))
```

```python
def search(nums, target):
    left, right = 0, len(nums) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        # Found the target
        if nums[mid] == target:
            return mid
        
        # Determine which side is sorted
        if nums[left] <= nums[mid]:  # Left side is sorted
            # Check if target is in the sorted left side
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        else:  # Right side is sorted
            # Check if target is in the sorted right side
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1
    
    return -1
nums = [1,3]
target = 1
print(search(nums,target))

```

```python
class Solution(object):
    def solveSudoku(self, board):
        self.rows = [set() for _ in range(9)]
        self.cols = [set() for _ in range(9)]
        self.boxes = [set() for _ in range(9)]
        
        for r in range(9):
            for c in range(9):
                if board[r][c] != '.':
                    num = board[r][c]
                    self.rows[r].add(num)
                    self.cols[c].add(num)
                    box_idx = (r // 3) * 3 + (c // 3)
                    self.boxes[box_idx].add(num)
        
        self.solve(board)
    
    def solve(self, board):
        # Find the empty cell with fewest possibilities (MCV heuristic)
        min_options = 10
        best_cell = None
        
        for r in range(9):
            for c in range(9):
                if board[r][c] == ".":
                    box_idx = (r // 3) * 3 + (c // 3)
                    # Count how many numbers are possible
                    options = 0
                    for num in "123456789":
                        if (num not in self.rows[r] and 
                            num not in self.cols[c] and 
                            num not in self.boxes[box_idx]):
                            options += 1
                    
                    if options < min_options:
                        min_options = options
                        best_cell = (r, c)
                        if options == 0:  # No valid options, dead end
                            return False
        
        # If no empty cells, puzzle is solved
        if best_cell is None:
            return True
        
        r, c = best_cell
        box_idx = (r // 3) * 3 + (c // 3)
        
        for num in "123456789":
            if (num not in self.rows[r] and 
                num not in self.cols[c] and 
                num not in self.boxes[box_idx]):
                
                board[r][c] = num
                self.rows[r].add(num)
                self.cols[c].add(num)
                self.boxes[box_idx].add(num)
                
                if self.solve(board):
                    return True
                
                board[r][c] = "."
                self.rows[r].remove(num)
                self.cols[c].remove(num)
                self.boxes[box_idx].remove(num)
        
        return False
```

```python
def countAndSay(n):
    s = "1"
    for k in range(n - 1):
        next_s = ""
        i = 0
        while i < len(s):
            count = 1
            while i + 1 < len(s) and s[i] == s[i+1]:
                count += 1
                i += 1
            next_s += str(count) + s[i]
            i += 1
        s = next_s
    return s

# Example
for i in range(1, 6):
    print(i, countAndSay(i))
```

```python
def firstMissingPositive(nums):
    n = len(nums)
    
    # Step 1: Place each number in its "correct" position
    # Number k should be at index k-1
    for i in range(n):
        while 1 <= nums[i] <= n and nums[nums[i] - 1] != nums[i]:
            # Swap nums[i] to its correct position
            correct_pos = nums[i] - 1

            nums[i], nums[correct_pos] = nums[correct_pos], nums[i]
            print(f"Now the list is {nums}, and index is {i}")
    print(nums)
    
    # Step 2: Find the first position where the number is wrong
    for i in range(n):
        if nums[i] != i + 1:
            return i + 1
    
    # If all positions are correct, answer is n+1
    return n + 1
print(firstMissingPositive([3,4,-1,1]))
```

```python
def trap(height):
 
    
    left, right = 0, len(height) - 1
    left_max, right_max = 0, 0
    rain = 0
    
    while left < right:
        if height[left] < height[right]:
            if height[left] >= left_max:
                left_max = height[left]
            else:
                rain += left_max - height[left]
            left += 1
        else:
            if height[right] >= right_max:
                right_max = height[right]
            else:
                rain += right_max - height[right]
            right -= 1
    
    return rain
```

```python
def groupAnagrams(strs):
    groups = {}
    
    for s in strs:
        key = ''.join(sorted(s))#put the char in '' together
        print(key)
        
        # Manual defaultdict behavior
        if key not in groups:
            groups[key] = []
        
        groups[key].append(s)
    
    return list(groups.values())
strs = ["eat","tea","tan","ate","nat","bat"]
print(groupAnagrams(strs))
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import DBSCAN
import numpy as np

def groupAnagrams_Embeddings(strs):
    # Load pretrained model
    model = SentenceTransformer('all-MiniLM-L6-v2')
    
    # Get embeddings for each string
    embeddings = model.encode(strs)
    
    # Cluster using DBSCAN
    clustering = DBSCAN(eps=0.5, min_samples=1, metric='cosine')
    labels = clustering.fit_predict(embeddings)
    
    # Group by cluster labels
    groups = {}
    for i, label in enumerate(labels):
        if label not in groups:
            groups[label] = []
        groups[label].append(strs[i])
    
    return list(groups.values())

# Test
print(groupAnagrams_Embeddings(["eat","tea","tan","ate","nat","bat"]))
```
