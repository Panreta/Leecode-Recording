# 76. Minimum Window Substring

```python
def minWindow(s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        ## CC
        if len(s) < len(t):
            return ""

        need,window = {},{}
        for char in t:
            need[char] = need.get(char,0) + 1

        
        left,right,valid = 0,0,0

        start = 0
        length = float('inf')
        while right < len(s):
            char = s[right]
            right += 1# proceed to the right

            if char in need:
                window[char] = window.get(char,0) + 1
                if window[char] == need[char]:#fulfill the requirement for one
                    valid += 1

            while valid == len(need): # fulfill for all char in t, but may be local minimum
            # update the length first, like min function
                if right - left < length:
                    start = left
                    length = right - left
                delete = s[left]
                left += 1

                #exploration, see the rest of the string
                if delete in need: 
                    if window[delete] == need[delete]:
                        valid -= 1
                    window[delete] -= 1

        return "" if length == float('inf') else s[start: start + length]

s,t = "ADOBECODEBANC","ABC"
print(minWindow(s,t))
```

# 567. Permutation in String

```python
def checkInclusion(s1, s2):
  
        need,window = {},{}
        # #char in s1
        for char in s1:
            need[char] = need.get(char,0) + 1

        left,right,valid = 0,0,0
        while right < len(s2):
            char = s2[right]
            right += 1

            if char in need:
                window[char] = window.get(char,0) + 1
                if window[char] == need[char]:
                    valid += 1

            while right - left >= len(s1):
                if valid == len(need):
                    return True

                delete = s2[left]
                left += 1
                if delete in need:
                    if window[delete] == need[delete]:
                        valid -= 1
                    window[delete] -= 1
        return False

```

# 3.Longest Substring Without Repeating Characters

```python
def lengthOfLongestSubstring(s):
    if len(s) == 0:
        return 0
    
    window = {}
    l,r = 0,0
    res = 0
    while r < len(s):
        char = s[r]
        r += 1
        window[char] = window.get(char,0) + 1

        if window[char] > 1:
            delete = s[l]
            l += 1
            window[char] = window.get(delete,0) - 1

        res = max(res,r - l)
    return res


        

                    
       
```
