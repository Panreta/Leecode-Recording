Given a string s, find the length of the longest substring without duplicate characters.

```python
def lengthOfLongestSubstring(s):
    if len(s) == 0:
        return 0
    
    window = {}
    l,r = 0,0
    res = 0
    while r < len(s):
        char = s[r]
        r += 1
        window[char] = window.get(char,0) + 1

        while window[char] > 1:
            delete = s[l]
            l += 1
            window[delete] = window.get(delete,0) - 1

        res = max(res,r - l)
    return res
```

# 567. Permutation in String

```python
def checkInclusion(s1, s2):
  
        need,window = {},{}
        # #char in s1
        for char in s1:
            need[char] = need.get(char,0) + 1

        left,right,valid = 0,0,0
        while right < len(s2):
            char = s2[right]
            right += 1

            if char in need:
                window[char] = window.get(char,0) + 1
                if window[char] == need[char]:
                    valid += 1

            while right - left >= len(s1):
                if valid == len(need):
                    return True

                delete = s2[left]
                left += 1
                if delete in need:
                    if window[delete] == need[delete]:
                        valid -= 1
                    window[delete] -= 1
        return False

```

209. Minimum Size Subarray Sum

```python
def minSubArrayLen(self, target, nums):
        """
        :type target: int
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)
        l = 0
        total = 0
        result = float('inf')

        for r in range(n):
            total += nums[r]
            while total >= target:
                result = min(result, r - l + 1)
                total -= nums[l]
                l += 1

        return 0 if result == float('inf') else result
```

30. Substring with Concatenation of All Words

```python
from collections import Counter

def findSubstring(s,words):
    word_len = len(words[0])
    word_count = len(words)
    window_len = word_len * word_count
    need = Counter(words)
    result = []

    for i in range(len(s) - window_len + 1):
        seen = {}
        j = 0
        while j < word_count:
            word = s[i + j * word_len: i + (j + 1) * word_len]
            if word not in need:
                break
            seen[word] = seen.get(word,0) + 1
            if seen[word] > need[word]:
                break
            j += 1
        if j == word_count:
            result.append(i)

    return result

s = "barfoothefoobarman"
words = ["foo","bar"]
print(Counter(words))


```

```python
from collections import Counter
def findSubstring(s, words):

    word_len   = len(words[0])
    n          = len(s)
    need       = Counter(words)
    result     = []

    # only word_len distinct starting offsets matter
    for offset in range(word_len):
        seen    = {}
        matched = 0
        l       = offset

        for r in range(offset, n - word_len + 1, word_len):
            word = s[r : r + word_len]

            if word in need:
                seen[word] = seen.get(word, 0) + 1
                if seen[word] == need[word]:
                    matched += 1

                # window too wide — shrink from left
                while seen[word] > need[word]:
                    left_word = s[l : l + word_len]
                    if seen[left_word] == need[left_word]:
                        matched -= 1
                    seen[left_word] -= 1
                    l += word_len

                if matched == len(need):
                    result.append(l)

            else:
                # invalid word — reset window
                seen    = {}
                matched = 0
                l       = r + word_len

    return result
    
```
