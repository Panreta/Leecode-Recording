909. Snakes and Ladders

```python
from collections import deque
def snakesAndLadders(board):
    """
    :type board: List[List[int]]
    :rtype: int
    """
    n = len(board)

    def num_to_pos(num):# given num, return the element position, shouldn't be hard to find
        num -= 1
        row,col = divmod(num,n)
        if row % 2 == 1: # odd row
            col = n - 1 - col
        return n - 1 - row, col

    visited = set()# record shortcut
    q = deque([(1,0)])

    while q:
        curr, moves = q.popleft()
        for dice in range(1,7):
            nxt = curr + dice
            if nxt > n * n:
                break
            r, c = num_to_pos(nxt)
            if board[r][c] != -1:
                nxt = board[r][c]
            if nxt == n * n:
                return moves + 1
            if nxt not in visited:
                visited.add(nxt)
                q.append((nxt,moves + 1))
    return -1
```

433. Minimum Genetic Mutation

```python
from collections import deque
def minMutation(startGene, endGene, bank):
    """
    :type startGene: str
    :type endGene: str
    :type bank: List[str]
    :rtype: int
    """
    bank = set(bank)
    if endGene not in bank:
        return -1

    visited = {startGene} # only need set actually, just make it simple to use hash set
    q = deque([(startGene,0)])

    while q:
        gene,move = q.popleft()
        if gene == endGene:
            return move

        for i in range(8):
            for c in "AGTC":
                mutated = gene[:i] + c + gene[i + 1:]
                if mutated in bank and mutated not in visited:
                    visited.add(mutated)
                    q.append((mutated,move + 1))

    return -1
```

```python
def ladderLength(beginWord, endWord, wordList):
        """
        :type beginWord: str
        :type endWord: str
        :type wordList: List[str]
        :rtype: int
        """
        from collections import deque
        def howDifferent(s,t):
            count = 0
            for i in range(len(s)):
                if s[i] != t[i]:
                    count += 1
            return count == 1
        if endWord not in wordList:
            return 0

        visited = {beginWord}
        q = deque([(beginWord,1)])

        while q:
            state,move = q.popleft()
            if state == endWord:
                return move 
            for word in wordList:
                if howDifferent(state,word) and word not in visited:
                    visited.add(word)
                    q.append((word,move + 1))

        return 0
```
