130. Surrounded Regions

```python
def solve(self, board):
        """
        :type board: List[List[str]]
        :rtype: None Do not return anything, modify board in-place instead.
        """
        m,n = len(board),len(board[0])
        
        def dfs(i,j):
            if i < 0 or i >=m or j < 0 or j >= n or board[i][j] != 'O':
                return
            board[i][j] = 'X'
            dfs(i - 1,j)
            dfs(i + 1,j)
            dfs(i,j - 1)
            dfs(i,j + 1)

        def dfs1(i,j):
            if i < 0 or i >=m or j < 0 or j >= n or board[i][j] != 'O':
                return
            board[i][j] = 'P'
            dfs1(i - 1,j)
            dfs1(i + 1,j)
            dfs1(i,j - 1)
            dfs1(i,j + 1)

        # Boundary
        for i in range(m):
            dfs1(i,0)
            dfs1(i,n - 1)
        for j in range(n):
            dfs1(0,j)
            dfs1(m - 1,j)

        for i in range(m):
            for j in range(n):
                if board[i][j] == 'O':
                    dfs(i,j)

        for i in range(m):
            for j in range(n):
                if board[i][j] == 'P':
                    board[i][j] = 'O'
        return board
```

Easier way:

```python
def solve(self, board):
    m, n = len(board), len(board[0])

    def dfs(i, j, mark):
        if i < 0 or i >= m or j < 0 or j >= n or board[i][j] != 'O':
            return
        board[i][j] = mark
        dfs(i-1, j, mark)
        dfs(i+1, j, mark)
        dfs(i, j-1, mark)
        dfs(i, j+1, mark)

    # mark border-connected O's as 'P'
    for i in range(m):
        dfs(i, 0,   'P')
        dfs(i, n-1, 'P')
    for j in range(n):
        dfs(0,   j, 'P')
        dfs(m-1, j, 'P')

    # flip remaining O's to X, restore P's to O
    for i in range(m):
        for j in range(n):
            if board[i][j] == 'O': board[i][j] = 'X'
            if board[i][j] == 'P': board[i][j] = 'O'
```

```python
def cloneGraph(self, node):
        """
        :type node: Node
        :rtype: Node
        """
        if not node:
            return None

        mapping = {}

        def dfs(node):
            if node in mapping:
                return mapping[node]

            clone = Node(node.val)
            mapping[node] = clone

            for neighbor in node.neighbors:
                clone.neighbors.append(dfs(neighbor))
            return clone
        return dfs(node)
```
