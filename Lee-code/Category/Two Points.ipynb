# 21. Merge Two Sorted Lists

```python
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def list_to_linked_list(lst):
    head = ListNode(lst[0])
    current = head
    for value in lst[1:]:
        current.next = ListNode(value)
        current = current.next
    return head

def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=' ')
        current = current.next
    print()

```

# 86. Partition List

```python
def partition(head, x):
        before = before_head = ListNode()
        after = after_head = ListNode()
        cur = head
        while cur:
            if cur.val < x:
                before.next = cur
                before = before.next
            else:
                after.next = cur
                after = after.next
            cur = cur.next
        after.next = None
        before.next = after_head.next
        return before_head.next
```

# 19. Remove Nth Node From End of List

```python
def removeNthFromEnd(head,n):
    # count how many do you have
    count = 0
    cur = head
    while cur:
        count += 1
        cur = cur.next
    count = count - n  # from start

    dummy = ListNode()
    dummy.next = head
    cur = dummy
    # want right on the parent of cancelled one
    while count and cur:
        cur = cur.next
        count -= 1
    # achieve the node
    cur.next = cur.next.next
    return dummy.next


head = [1,2,3,4,5]
head = list_to_linked_list(head)
n = 2
print_linked_list(removeNthFromEnd(head,n))
```

# 142. Linked List Cycle II

distance from head  to cycle start = a

distance from cycle start to meeting point = b

distance from meeting point back to cycle start = c

When they meet inside the cycle:

slow travelled: a + b

fast travelled: a + b + c + b = a + 2b + c

Since fast = 2 × slow:

a + 2b + c ≡ 2(a + b)

a + 2b + c ≡ 2a + 2b

c ≡ a

```python
def detectCycle(head):
    # dream: two nodes can meet at the beginning of the cycle
    fast = slow = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break
    else:
        return None
    
    slow = head
    while slow != fast:
        slow = slow.next
        fast = fast.next

    return slow
```

# 160. Intersection of Two Linked Lists

Another trick in math:

A+C: path for a

B+C: path for b

WLOG: A<B: b has B-A length to go, B-A+A length to get to the start of the circle-start, a also need to walk B length to get to circle-start, so when they meet, it will be circle-start.

```python
def getIntersectionNode(headA, headB):
    a,b = headA, headB
    # the one jumps is the one with smaller branch.
    while a!=b:
        a = a.next if a else headB
        b = b.next if b else headA

    return a
```
