# 56. Merge Intervals

```python
def merge(intervals):
    intervals.sort(key=lambda x: x[0])
    merged = [intervals[0]]

    for start, end in intervals[1:]:
        if start <= merged[-1][1]:
            merged[-1][1] = max(merged[-1][1], end)
        else:
            merged.append([start, end])

    return merged
```

# 57. Insert Interval

You put one into the intervals, and just the same thing

```python
def insert(intervals, newInterval):

        intervals.append(newInterval)
        intervals.sort(key = lambda x: x[0])
        merged = [intervals[0]]

        for start, end in intervals[1:]:
            if start <= merged[-1][1]:
                merged[-1][1] = max(merged[-1][1],end)

            else:
                merged.append([start,end])

        return merged
```

# 252. Meeting Rooms:

Given an array of meeting time intervals where intervals[i] = [starti, endi], determine if a person could attend all meetings.

```python
def canAttendMeetings(intervals):
    
        if not intervals:
            return True
        intervals.sort(key = lambda x: x[0])

        merged = [intervals[0]]
        for start, end in intervals[1:]:
            if start < merged[-1][1]:
                return False

            else:
                merged.append([start,end])

        return True
```

# 253. Meeting room

Given an array of meeting time intervals intervals where intervals[i] = [starti, endi], return the minimum number of conference rooms required.

There is an implicit maximum behind the whole problem, imagine a three-overlapping list with 4-o list, the room will be 3, and not change until j move to the third list of 4-o. In another word, if there is no higher-o list than 3-o, 3 will never update.

```python
def minMeetingRooms(intervals):
    starts = sorted(x[0] for x in intervals)
    ends = sorted(x[1] for x in intervals)

    rooms = 0
    j = 0
    for i in range(len(starts)):
        if starts[i] < ends[j]:
            rooms += 1  # new meeting starts before any room frees up
        else:
            j += 1      # a room freed up, reuse it
    return rooms
```

```python
import heapq

# heapq is a min-heap by default
heap = []

# 1. Push
heapq.heappush(heap, 5)
heapq.heappush(heap, 2)
heapq.heappush(heap, 8)
heapq.heappush(heap, 1)
print(heap)  # [1, 2, 8, 5] — stored as a list, smallest at index 0

# 2. Pop (always removes the smallest)
print(heapq.heappop(heap))  # 1
print(heapq.heappop(heap))  # 2

# 3. Peek without popping
print(heap[0])  # just index into the list

# 4. Heapify — turn an existing list into a heap in O(n)
nums = [5, 2, 8, 1]
heapq.heapify(nums)
print(nums)  # [1, 2, 8, 5]

# 5. heapreplace — pop min and push new in ONE operation (faster than pop+push)
heapq.heapreplace(heap, 10)

# 6. nlargest / nsmallest
nums = [5, 2, 8, 1, 9, 3]
print(heapq.nsmallest(3, nums))  # [1, 2, 3]
print(heapq.nlargest(3, nums))   # [9, 8, 5]

# 7. Max-heap trick — negate the values
heap = []
heapq.heappush(heap, -5)
heapq.heappush(heap, -2)
heapq.heappush(heap, -8)
print(-heapq.heappop(heap))  # 8 (largest)

# 8. Heap of tuples — compares by first element
heap = []
heapq.heappush(heap, (5, "task A"))
heapq.heappush(heap, (1, "task C"))
heapq.heappush(heap, (3, "task B"))
print(heapq.heappop(heap))  # (1, 'task C')
```

```python
import heapq

def minMeetingRooms(intervals):
    if not intervals:
        return 0
    intervals.sort(key=lambda x: x[0])
    heap = []  # tracks end times of active rooms

    for start, end in intervals:
        if heap and heap[0] <= start:
            heapq.heapreplace(heap, end)  # recycle earliest-ending room
        else:
            heapq.heappush(heap, end)     # need a new room

    return len(heap)
```

# 435. Non-overlapping Intervals

```python
def eraseOverlapIntervals(intervals):
   
        intervals.sort(key=lambda x: x[1])
        keep = 1
        cur_end = intervals[0][1]

        for start,end in intervals[1:]:
            if start >= cur_end:
                keep += 1
                cur_end = end
        return len(intervals) - keep
        
```
