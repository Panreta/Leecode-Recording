Given a 1-indexed array of integers numbers that is already sorted in non-decreasing order, find two numbers such that they add up to a specific target number. Let these two numbers be numbers[index1] and numbers[index2] where 1 <= index1 < index2 <= numbers.length.

Return the indices of the two numbers index1 and index2, each incremented by one, as an integer array [index1, index2] of length 2.

The tests are generated such that there is exactly one solution. You may not use the same element twice.

Your solution must use only constant extra space.

If we ignore everything about the sorted array, we can still use hash table to make the result.

```python
def twoSum(numbers, target):
    """
    :type numbers: List[int]
    :type target: int
    :rtype: List[int]
    """
    dict = {}
    for i,j in enumerate(numbers):
        if target - j in dict:
            return[dict[target-j]+1,i+1]
        dict[j] = i
```

However, we wanna the specific answer with 2 pointers, so the next solution comes:

```python
def twoSum(numbers, target):

    slow, fast = 0, len(numbers) - 1

    while numbers[slow] + numbers[fast] != target:
        if numbers[slow] + numbers[fast] < target:
            slow += 1
        else:
            fast -= 1

    return [slow+1,fast+1]
```

You can tell, the problem is using every char as a center and distinguish the left side and right side, whether if they are equal to each other. Also, I believe this is like a slide window.

```python
def longestPalindrome(s):
        """
        :type s: str
        :rtype: str
        """
        if len(s) == 1:
            return s
        result = ''

        for center in range(len(s) - 1):
            # odd length
            j = 1
            while center - j >= 0 and center + j <= len(s) - 1:
                if s[center - j] == s[center + j]:
                    j += 1
                else:
                    break
            if len(s[center - j + 1:center + j]) > len(result):
                result = s[center - j+1:center + j]
                
            # even length
            slow, fast = center,center + 1
            j = 0
            while slow - j >=0 and fast + j <= len(s) - 1:
                if s[slow - j] == s[fast + j]:
                    j += 1
                else:
                    break
            if len(s[slow - j + 1:fast + j]) > len(result):
                result = s[slow - j + 1:fast + j]
            

        return result
```

Faster and modern one:

```python

def longestPalindrome(s):
    def expand(l, r):
        while l >= 0 and r < len(s) and s[l] == s[r]:
            l -= 1
            r += 1
        return s[l + 1 : r]   # l,r are one step too far after the loop

    result = ''
    for i in range(len(s)):
        odd  = expand(i, i)       # odd  length: single center
        even = expand(i, i + 1)   # even length: two centers
        result = max(result, odd, even, key=len)

    return result
```
