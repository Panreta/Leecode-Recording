# 88. Merge Sorted Array

Must do from the back cuz you don't wanna push all the elements in nums1 one back.

Choosing max + start from the empty part

```python
def merge(nums1, m, nums2, n):
        """
        :type nums1: List[int]
        :type m: int
        :type nums2: List[int]
        :type n: int
        :rtype: None Do not return anything, modify nums1 in-place instead.
        """
        i,j,final_in_nums1 = m - 1, n - 1, m + n - 1
        
        while i >= 0 and j >= 0:
            if nums1[i] > nums2[j]:# nums1[i] is the largest one for now, >= is also ok
                nums1[final_in_nums1] = nums1[i]
                i -= 1
            else: 
                nums1[final_in_nums1] = nums2[j]
                j -= 1
            final_in_nums1 -= 1

        # when i>=0, which means the element in nums1 already in their places
        # when j >=0, which means numbers should be put into nums1
        while j >= 0:
            nums1[final_in_nums1] = nums2[j]
            j -= 1
            final_in_nums1 -= 1

                  
        

# Test Cases
print(merge([1, 2, 3, 0, 0, 0], 3, [2, 5, 6], 3))
```

26. Remove Duplicates from Sorted Array

```python
def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        i,j = 0,1
        while j < len(nums):
            if nums[i] == nums[j]:
                 continue
            else:# only add when they are not the same
                nums[i + 1] = nums[j]
                i += 1
            j += 1

        return i + 1
```

# 169. Majority Element

My code, O(nlogn) for average

```python
def majorityElement(nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        frequency = {}
        for num in nums:
            frequency[num] = frequency.get(num,0) + 1
        sort_frequency = sorted(frequency.items(),key = lambda x: x[1])
        return sort_frequency[-1][0]

print(majorityElement([2,2,1,1,1,2,2]))
```

```python
def majorityElement(nums):
    t = nums[0]
    say = 1
    for i in range(1,len(nums)):
        if nums[i] == t:
            say += 1
        elif say == 0:# only when this new element get more than the used one
            t = nums[i] 
        else:
            say -= 1
    return t
```

```python
def rotate(self, nums, k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        length = len(nums)
        k = k % length
        nums[:] = nums[length - k:] + nums[:length - k]
```

121. Best Time to Buy and Sell Stock

Only sell once, so must wait for the biggest difference.

```python
def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """
        n = len(prices)
        min_value = float('inf')
        max_benefit = 0

        for price in prices:
            if price < min_value:
                min_value = price
            else:# the price is larger
                max_benefit = max(max_benefit,price - min_value)
        return max_benefit
```

123. Best Time to Buy and Sell Stock III

Hold at most 2 stock each time. Imagine every time holding the stocks is the state, and whether to sell is an action.

```python
def maxProfit(prices):
    """
    :type prices: List[int]
    :rtype: int
    """
    if not prices:
        return 0
    
    # Track 4 states: after 1st buy, 1st sell, 2nd buy, 2nd sell
    buy1 = float('-inf')# if you don't buy, the penalty is large
    sell1 = 0
    buy2 = float('-inf')
    sell2 = 0
    
    for price in prices:
        # First transaction, every step, just compare if we need to make a move
        buy1 = max(buy1, -price)           # Buy at current price
        sell1 = max(sell1, buy1 + price)   # Sell after first buy, price_now + (-price_bought)
        
        # Second transaction
        buy2 = max(buy2, sell1 - price)    # Buy again after first sell
        sell2 = max(sell2, buy2 + price)   # Sell after second buy
    
    return sell2
```

```python
# At most k transactions
def maxProfit(prices, k):
    buy  = [float('-inf')] * k
    sell = [0] * k

    for price in prices:
        for i in range(k):
            buy[i]  = max(buy[i],  (sell[i-1] if i > 0 else 0) - price)
            sell[i] = max(sell[i], buy[i] + price)

    return sell[-1]
```

55. Jump Game

```python
def canJump(nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        fastest = 0

        for i in range(len(nums)):
            if i > fastest:
                return False
            fastest = max(fastest, i + nums[i])

        return True
        
print(canJump([2,3,1,1,4]))
```

45. Jump Game II

```python
def jump(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        jump, fastest, end = 0,0,0
        for i in range(len(nums) - 1):# you don't have to jump at late
            fastest = max(fastest,i + nums[i])
            if i == end:
                jump += 1
                end = fastest

        return jump
```

274. H-Index

We can tell, actually sort can bring the optimization structure out(min and max are so obvious)

```python
def hIndex(self, citations):
        """
        :type citations: List[int]
        :rtype: int
        """
        citations.sort(reverse = True)
        h = 0

        for idx,citation in enumerate(citations):
            if citation >= idx + 1:
                h = idx + 1
            else:
                break

        return h
```

So the formula is:$$ answer[i] = \prod_{j=1,j\neq i}^{n}nums[j]=\prod_{j=1}^{i-1}nums[j]\times \prod_{j=i+1}^{n}nums[j]=preProduct * suffixProduct$$

```python
def productExceptSelf(nums):
    n = len(nums)
    answer = [1] * n

    # Step 1: fill answer with prefix products
    prefix = 1
    for i in range(n):
        answer[i] = prefix
        prefix *= nums[i]
    print(f"answer is {answer}")

    # Step 2: multiply with suffix products from the right
    suffix = 1
    for i in range(n - 1, -1, -1):
        answer[i] *= suffix
        suffix *= nums[i]
    

    return answer
print(productExceptSelf([1,2,3,4]))
```

```python
def canCompleteCircuit(gas, cost):
        """
        :type gas: List[int]
        :type cost: List[int]
        :rtype: int
        """
        if sum(gas) < sum(cost):
            return -1

        remain = 0
        start = 0
        for i in range(len(gas)):
            remain += gas[i] - cost[i]
            if remain < 0:
                start = i + 1
                remain = 0
        return start
```
