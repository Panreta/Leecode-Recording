# List <-> String

```python
def reverseWords(input):
    # Split the string by spaces into a list of words
    inputWords = input.split(",")
    print(f"list: {inputWords}")

    # inputWords[-1::-1] — three slice parameters:
    # 1st parameter -1: start from the last element
    # 2nd parameter (empty): move until the beginning of the list
    # 3rd parameter -1: step backwards (move one position left each time)
    inputWords = inputWords[-1::-1]

    # Re-join the words with commas
    output = ','.join(inputWords)
    print(f"list back to be string: {output}")


    return output

input = 'Hello,World'
print(reverseWords(input))
```

# Bitwise Operator

```python
a = 60            # 60 = 0011 1100
b = 13            # 13 = 0000 1101
c = 0

c = a & b         # 12 = 0000 1100  (AND)
print("1 - value of c:", c)

c = a | b         # 61 = 0011 1101  (OR)
print("2 - value of c:", c)

c = a ^ b         # 49 = 0011 0001  (XOR)
print("3 - value of c:", c)

c = ~a            # -61 = 1100 0011 (NOT)
print("4 - value of c:", c)

c = a << 2        # 240 = 1111 0000 (left shift)
print("5 - value of c:", c)

c = a >> 2        # 15 = 0000 1111  (right shift)
print("6 - value of c:", c)
```

```python
# keys only — values default to None
dict.fromkeys(["a", "b", "c"])
# {'a': None, 'b': None, 'c': None}

# with a default value
dict.fromkeys(["a", "b", "c"], 0)
# {'a': 0, 'b': 0, 'c': 0}

# from a string (each character becomes a key)
dict.fromkeys("hello")
# {'h': None, 'e': None, 'l': None, 'o': None}
```

```python
text = "hello world"
counts = {}
for char in text:
    counts[char] = counts.get(char, 0) + 1
# {'h': 1, 'e': 1, 'l': 3, 'o': 2, ...}
```

```python
d = {"a": 1, "b": 2, "c": 3}

d.items()  # dict_items([('a', 1), ('b', 2), ('c', 3)])

# most common use — loop over both key and value
for key, value in d.items():
    print(key, value)
# a 1
# b 2
# c 3
```

```python
d = {"a": 1, "b": 2, "c": 3}

d.keys()   # dict_keys(['a', 'b', 'c'])
d.values() # dict_values([1, 2, 3])

for key in d.keys():
    print(key)   # a, b, c

# check membership
"a" in d.keys()  # True
```

```python
class MyNumbers:
    def __iter__(self):
        self.a = 1
        return self        # ✅ indented inside __iter__

    def __next__(self):
        x = self.a
        self.a += 1
        return x

myclass = MyNumbers()
myiter = iter(myclass)

print(next(myiter))  # 1
print(next(myiter))  # 2
print(next(myiter))  # 3
print(next(myiter))  # 4
print(next(myiter))  # 5
```

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

# create generator object
generator = countdown(5)

# get values by iterating the generator
print(next(generator))  # 5
print(next(generator))  # 4
print(next(generator))  # 3

# use for loop to iterate the remaining values
for value in generator:
    print(value)  # 2 1
```

# Decorator function: Give out more info

```python
def logger(original_function):
    def wrapper(*args, **kwargs):
        print(f"calling {original_function.__name__} with {args}")  # before
        result = original_function(*args, **kwargs)
        print(f"returned: {result}")                                 # after
        return result
    return wrapper

@logger
def add(a, b):
    return a + b

add(3, 5)
# calling add with (3, 5)
# returned: 8
```

Para

```python
def repeat(num_times):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(num_times):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(3)
def say_hello():
    print("Hello!")

say_hello()
```

# Inherent

```python
class people:
    def __init__(self, n, a, w, **kwargs):  # accept extra kwargs and pass them up
        self.name = n
        self.age = a
        self.__weight = w
        super().__init__(**kwargs)

    def speak(self):
        print("%s says: I am %d years old." % (self.name, self.age))


class student(people):
    grade = ''

    def __init__(self, n, a, w, g, **kwargs):
        self.grade = g
        super().__init__(n, a, w, **kwargs)  # passes remaining kwargs up the chain

    def speak(self):
        print("%s says: I am %d years old, in grade %d" % (self.name, self.age, self.grade))


class speaker:
    def __init__(self, n, t, **kwargs):
        self.name = n
        self.topic = t
        super().__init__(**kwargs)

    def speak(self):
        print("My name is %s, I am a speaker, my topic is %s" % (self.name, self.topic))


class sample(speaker, student):
    def __init__(self, n, a, w, g, t):
        super().__init__(n=n, a=a, w=w, g=g, t=t)  # ✅ one single super() call


s = sample('ken', 10, 60, 3, 'Python')
s.speak()  # My name is ken, I am a speaker, my topic is Python
```

# public or private method

```python
class Site:
    def __init__(self, name, url):
        self.name = name       # public
        self.__url = url   # private
 
    def who(self):
        print('name  : ', self.name)
        print('url : ', self.__url)
 
    def __foo(self):          # 私有方法
        print('这是私有方法')
 
    def foo(self):            # 公共方法
        print('这是公共方法')
        self.__foo()
```
