```python
def permute(nums):
    visit = set(nums)
    path = []
    ans = []

    def dfs(num):
        visit.remove(num)
        path.append(num)
        if len(visit) == 0:
            ans.append(path[:])
        else:
            for j in nums:
                if j in visit:
                    dfs(j)
        visit.add(num)
        path.pop()
    

    for num in nums:
        dfs(num)
    return ans

print(permute([1,2,3]))
```

# 47. Permutation with duplicated elements

```python
def permuteUnique(nums):
    res = []
    used = [False] * len(nums)
    nums.sort()

    def backtrack(path):
        if len(path) == len(nums):
            res.append(path[:])
            return
        for i in range(len(nums)):
            if used[i]:
                continue
            if i > 0 and nums[i] == nums[i-1] and not used[i-1]:
                continue
            used[i] = True
            path.append(nums[i])
            backtrack(path)
            path.pop()
            used[i] = False

    backtrack([])
    return res
```

```python
def nextPermutation(nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        n = len(nums)

        # first rightmost i s.t. nums[i] < nums[i + 1]
        i = n - 2
        while i >= 0 and nums[i] >= nums[i+1]:
            i -= 1

        if i >= 0:
            # rightmost j s.t.nums[j] > nums[i]
            j = n - 1
            while nums[j] <= nums[i]:
                j -= 1
            # swap
            nums[i],nums[j] = nums[j], nums[i]

        nums[i+1:] = nums[i + 1:][::-1] 
```

```python
def getPermutation(n, k):
    from math import factorial
    nums = list(range(1,n+1))
    k -= 1

    result = []
    for i in range(n,0,-1):
        fact = factorial(i - 1)
        idx = k // fact
        result.append(str(nums[idx]))
        nums.pop(idx)
        k %= fact
    return "".join(result)
```

# 996. Number of Squareful Arrays

```python
from collections import Counter
import math
class Solution(object):
    def numSquarefulPerms(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        def is_perfect_square(n):
            r = int(math.sqrt(n))
            return r * r == n

        def can_follow(x,y):
            return is_perfect_square(x + y)

        count = Counter(nums)
        unique = list(count.keys())

        def backtrack(last,remaining):
            if remaining == 0:
                return 1
            total = 0

            for nxt in unique:
                if count[nxt] > 0 and (last is None or can_follow(last, nxt)):
                    count[nxt] -= 1
                    total += backtrack(nxt, remaining - 1)
                    count[nxt] += 1

            return total

        return backtrack(None, len(nums))



    
```
