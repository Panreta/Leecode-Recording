# 88. Merge Sorted Array

Must do from the back cuz you don't wanna push all the elements in nums1 one back

```python
def merge(nums1, m, nums2, n):
        """
        :type nums1: List[int]
        :type m: int
        :type nums2: List[int]
        :type n: int
        :rtype: None Do not return anything, modify nums1 in-place instead.
        """
        i,j,final_in_nums1 = m - 1, n - 1, m + n - 1
        
        while i >= 0 and j >= 0:
            if nums1[i] > nums2[j]:# nums1[i] is the largest one for now, >= is also ok
                nums1[final_in_nums1] = nums1[i]
                i -= 1
            else: 
                nums1[final_in_nums1] = nums2[j]
                j -= 1
            final_in_nums1 -= 1

        # when i>=0, which means the element in nums1 already in their places
        # when j >=0, which means numbers should be put into nums1
        while j >= 0:
            nums1[final_in_nums1] = nums2[j]
            j -= 1
            final_in_nums1 -= 1

                  
        

# Test Cases
print(merge([1, 2, 3, 0, 0, 0], 3, [2, 5, 6], 3))
```

```python
def removeElement(nums,val):
    if not nums: # must have a list fisrt
        return 0
    
    i,j = 0, len(nums) - 1

    while i <= j: # since don't mind if i == j, so for len(nums) == 1 also works. Remind of i == j works
        if nums[i] == val: # option 1: nums[i] = val; option 2: nums[j + 1] = val.
            nums[i] = nums[j]
            j -= 1

        else: # only pluc when nums[i] != val
            i += 1
    return i,nums

print(removeElement([3,2,2,3],3))
```

```python
class Solution(object):
    def removeElement(self, nums, val):
        """
        :type nums: List[int]
        :type val: int
        :rtype: int
        """
        ## cc\
        if not nums:
            return 0
        if len(nums) == 1:
            return 0 if nums[0] == val else 1

        i,j = 0, len(nums) - 1
        while i < j :
            if nums[i] == val:
                while i < j and nums[j] == val:
                    j -= 1
            
                
                nums[i] = nums[j]
                nums[j] = val
                i += 1

            else:
                i += 1

        if nums[j] != val:
            return j + 1
        else:
            return i

        
```

26. Remove Duplicates from Sorted Array

```python
def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        i,j = 0,1
        while j < len(nums):
            if nums[i] == nums[j]:
                 continue
            else:# only add when they are not the same
                nums[i + 1] = nums[j]
                i += 1
            j += 1

        return i + 1
```
