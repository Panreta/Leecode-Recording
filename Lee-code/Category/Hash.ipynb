Apart from backtrack using to find power-set, it uses to find the subarrays.

```python
def subarraySum(nums, k):
    count = 0
    prefix_sum = 0
    freq = {0: 1}  # 前缀和 → 出现次数

    for num in nums:
        prefix_sum += num
        # 如果 prefix_sum - k 存在，说明中间那段子数组的和 == k
        count += freq.get(prefix_sum - k, 0)
        print(f"num: {num}, prefix_sum: {prefix_sum}, count: {count}, freq: {freq}")
        print(freq.get(prefix_sum - k, 0))
        freq[prefix_sum] = freq.get(prefix_sum, 0) + 1

    return count

print(subarraySum([1, 1, 1], 2))  # 输出 2
```

383. Ransom Note

```python
from collections import Counter
def canConstruct(self, ransomNote, magazine):
    """
    :type ransomNote: str
    :type magazine: str
    :rtype: bool
    """
    need = Counter(ransomNote)
    material = Counter(magazine)
    for key in need:
        if key not in material:
            return False
        if need[key] > material[key]:
            return False
    return True
```

Notice the mapping and reverse_mapping:

```python
def isIsomorphic( s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        if len(s) != len(t):
            return False
        left_mapping, right_mapping = {}, {}
        count = 0

        for char in s:
            if char in left_mapping: # mapping
                if left_mapping[char] != t[count]:
                    return False

            if t[count] in right_mapping:  # reverse mapping 
                if right_mapping[t[count]] != char:
                    return False

            left_mapping[char] = t[count]
            right_mapping[t[count]] = char
            count += 1

        return True
```

290. Word Pattern   

```python
def wordPattern(pattern, s):
        """
        :type pattern: str
        :type s: str
        :rtype: bool
        """
        l = s.split()
        if len(l) != len(pattern):
            return False

        mapping = {}
        reverse_mapping = {}

        for i in range(len(pattern)):
            if pattern[i] in mapping and mapping[pattern[i]] != l[i]:
                return False
            if l[i] in reverse_mapping and reverse_mapping[l[i]] != pattern[i]:
                return False
            mapping[pattern[i]] = l[i]
            reverse_mapping[l[i]] = pattern[i]
        return True

pattern = "abba"
s = "dog cat cat dog"
print(wordPattern(pattern,s))
```

242. Valid Anagram

```python
def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        from collections import Counter
        if len(s) != len(t):
            return False
        c1 = Counter(s)
        c2 = Counter(t)

        for key in c1:
            if c1[key] != c2[key]:
                return False
        return True
```

49. Group Anagrams

```python
from collections import defaultdict
def groupAnagrams(strs):
    groups = defaultdict(list)# accept missing key, key -> list
    for s in strs:
        key = "".join(sorted(s))   # sort return a list
        groups[key].append(s)
    print(groups)
    return list(groups.values())

strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(groupAnagrams(strs))
```

202. Happy Number

More like a number processing than hash

```python
def isHappy(n):
        """
        :type n: int
        :rtype: bool
        """
        seen = set()

        while True:
            result = 0
            s = str(n)
            for char in s:
                result += int(char) ** 2
            if result in seen:
                return False
            elif result == 1:
                return True
            else:
                seen.add(result)
                n = result
```

```python
def containsNearbyDuplicate(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: bool
        """
        seen = set()
        for i,num in enumerate(nums):
            if num in seen:
                return True
            seen.add(num)
            if len(seen) > k:
                seen.remove(nums[i - k])
        return False
```

128. Longest Consecutive Sequence

```python
def longestConsecutive(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        num_set = set(nums)
        result = 0
        for num in num_set:
            if num - 1 not in num_set:# find the start of a sequence
                length = 0
                while num + length in num_set:
                    length += 1
                result = max(result,length)

        return result
```
