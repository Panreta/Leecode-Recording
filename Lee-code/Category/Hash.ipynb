# hashmap

Apart from backtrack using to find power-set, it uses to find the subarrays.

```python
def subarraySum(nums, k):
    count = 0
    prefix_sum = 0
    freq = {0: 1}  # 前缀和 → 出现次数

    for num in nums:
        prefix_sum += num
        # 如果 prefix_sum - k 存在，说明中间那段子数组的和 == k
        count += freq.get(prefix_sum - k, 0)
        print(f"num: {num}, prefix_sum: {prefix_sum}, count: {count}, freq: {freq}")
        print(freq.get(prefix_sum - k, 0))
        freq[prefix_sum] = freq.get(prefix_sum, 0) + 1

    return count

print(subarraySum([1, 1, 1], 2))  # 输出 2
```
