BFS:

```python
from collections import deque
def maxDepth(self, root):
    """
    :type root: Optional[TreeNode]
    :rtype: int
    """
    if not root:
        return 0

    depth = 0
    q = deque([root])

    while q:
        for _ in range(len(q)):
            node = q.popleft()
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
        depth += 1
    return depth
```

101. Symmetric Tree

Since the iterative func needs 2 elements, you can't only use one

```python
def isSymmetric(root):
        """
        :type root: Optional[TreeNode]
        :rtype: bool
        """
        def dfs(p,q):
            if not p and not q:
                return True
            if not p or not q:
                return False
            return (p.val == q.val and dfs(p.left,q.right) and dfs(p.right,q.left))
        return dfs(root.left,root.right)
```

# 105. Construct Binary Tree from Preorder and Inorder Traversal

Unique orders to determine a tree.

preorder + inorder✅

postorder + inorder✅

Find

```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: Optional[TreeNode]
        """
        if not preorder:
            return None

        root = TreeNode(preorder[0])
        mid = inorder.index(preorder[0])

        root.left = self.buildTree(preorder[1:mid+1],inorder[:mid])
        root.right = self.buildTree(preorder[mid+1:],inorder[mid+1:])

        return root
```

```python
def buildTree(self, inorder, postorder):
        """
        :type inorder: List[int]
        :type postorder: List[int]
        :rtype: Optional[TreeNode]
        """
        if not inorder:
            return None

        root = TreeNode(postorder[-1])
        mid = inorder.index(postorder[-1])

        root.left = self.buildTree(inorder[:mid],postorder[:mid])
        root.right = self.buildTree(inorder[mid+1:],postorder[mid:-1])

        return root
```

117. Populating Next Right Pointers in Each Node II

```python
from collections import deque
class Solution(object):
    def connect(self, root):
        """
        :type root: Node
        :rtype: Node
        """
        if not root:
            return root

        q = deque([root])
        while q:
            size = len(q)
            for i in range(size):
                node = q.popleft()
                if i < size - 1: # can't be if i < len(q) - 1 because q is changing
                    node.next = q[0]
                if node.left: q.append(node.left)
                if node.right: q.append(node.right)
        return root
```

```python
def flatten(self, root):
    """
    :type root: Optional[TreeNode]
    :rtype: None Do not return anything, modify root in-place instead.
    """
    def preorder(node):
        if not node:
            return None
        left = node.left
        right = node.right

        node.left = None
        node.right = left
        preorder(left)

        tail = node
        while tail.right:
            tail = tail.right
        tail.right = right
        preorder(right)
    preorder(root)
```

```python
def hasPathSum(self, root, targetSum):
        """
        :type root: Optional[TreeNode]
        :type targetSum: int
        :rtype: bool
        """
        if not root:
            return False
        targetSum -= root.val
        if not root.left and not root.right:
            return targetSum == 0
        return self.hasPathSum(root.left, targetSum) or self.hasPathSum(root.right, targetSum)
```

```python
from collections import deque
def hasPathSum(self, root, targetSum):
    """
    :type root: Optional[TreeNode]
    :type targetSum: int
    :rtype: bool
    """
    if not root:
        return False
    
    q = deque([(root,targetSum)])
    while q:
        node,remain = q.popleft()
        remain -= node.val
        if not node.left and not node.right and remain == 0:
            return True
        if node.left: q.append((node.left,remain))
        if node.right: q.append((node.right,remain))
    return False
```
