```python
# Step 1: where can I put this number in? (list,num) -> index
from bisect import bisect_left, bisect_right


arr = [1, 3,4,4, 5, 6,6,7, 9]

print(bisect_left(arr, 4))   # 2  →  [1, 3, 4, 4, 5, 7, 9]
print(bisect_right(arr, 4))  # 4  →  same result when value is NOT in list
print(arr) # the arr is maintained


# Step 2: Put number into the list
from bisect import insort

arr = [1, 3, 5, 7]
insort(arr, 4)
print(arr)  # [1, 3, 4, 5, 7]

# Step 3: Binary Search
from bisect import bisect_left

def binary_search(arr, target):
    i = bisect_left(arr, target)
    if i < len(arr) and arr[i] == target:# from left, it will replace the original entry
        return i
    return -1

arr = [1, 3, 5, 7, 9]
print(binary_search(arr, 5))   # 2
print(binary_search(arr, 6))   # -1
```

# How to sort one list by another

```python
names = ['Alice', 'Bob', 'Charlie', 'Diana']
scores = [85, 72, 95, 68]

# Sort by scores ascending
paired = sorted(zip(scores, names)) # zip: list,list -> [list_entry,list_entry,...]
# paired = [(68, 'Diana'), (72, 'Bob'), (85, 'Alice'), (95, 'Charlie')]

# Unzip back into two lists
sorted_scores, sorted_names = zip(*paired)
print(list(sorted_names))   # ['Diana', 'Bob', 'Alice', 'Charlie']
print(list(sorted_scores))  # [68, 72, 85, 95]
```

```python
# Method 2: using SQLite
import sqlite3

conn = sqlite3.connect(":memory:")  # in-memory database
cur = conn.cursor()

cur.execute("CREATE TABLE students (name TEXT, score INTEGER);")
cur.executemany("INSERT INTO students VALUES (?, ?);",
                [('Alice', 85), ('Bob', 72), ('Charlie', 95), ('Diana', 68)])

cur.execute("SELECT name, score FROM students ORDER BY score DESC;")
print(cur.fetchall())
# [('Charlie', 95), ('Alice', 85), ('Bob', 72), ('Diana', 68)]

conn.close()
```

```python
# Method 3: using pandas + SQL syntax
import pandas as pd

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'score': [85, 72, 95, 68]
})

# Pure pandas
sorted_df = df.sort_values('score', ascending=False)
print(sorted_df)

# Or query with SQL syntax via pandasql
# pip install pandasql
from pandasql import sqldf
result = sqldf("SELECT * FROM df ORDER BY score DESC", locals())
print(result)
```
