228. Summary Ranges

```python
def summaryRanges(nums):
        """
        :type nums: List[int]
        :rtype: List[str]
        """
        if len(nums) == 1:
            return [str(nums[0])]
        nums.append(0)
        result = []
        start = nums[0]

        for i in range(len(nums) - 1):
            
            if nums[i] + 1 == nums[i + 1]:
                continue
            else:
                if start == nums[i]:
                    result.append(str(start))
                    start = nums[i + 1]
                else:
                    result.append(str(start) + "->" + str(nums[i]))
                    start = nums[i + 1]
        return result
```
