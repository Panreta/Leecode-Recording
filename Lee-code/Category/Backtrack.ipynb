# 17. Letter Combinations of a Phone Number

```python
def letterCombinations(self, digits):
        """
        :type digits: str
        :rtype: List[str]
        """
        res = []
        digitToChar = {
            "2": "abc",
            "3": "def",
            "4": "ghi",
            "5": "jkl",
            "6": "mno",
            "7": "qprs",
            "8": "tuv",
            "9": "wxyz",
        }
        def backtrack(i,curPath):
            if len(curPath) == len(digits):# determine what to extract
                res.append(curPath)
                return

            for char in digitToChar[digits[i]]:# growing tree logic
                backtrack(i + 1, curPath + char)


        backtrack(0,"")
        return res
```

The sequence matters: [1,2] is the same as [2,1], so 46. Permutations is a different problem.


```python
def combine(n, k):
        result = []
        
        def backtrack(start, current):
            if len(current) == k:
                result.append(current[:])
                return
            
            for i in range(start, n + 1):
                current.append(i)# since we can't do backtrack(i + 1, current.append(i)), we can only add first and pop later
                backtrack(i + 1, current)
                current.pop()
        
        backtrack(1, [])
        return result
```

```python

def permute(nums):
    """
    :type nums: List[int]
    :rtype: List[List[int]]
    """
    vis = set(nums)
    ans = []
    res = []
    def dfs(i):
        res.append(i)
        vis.remove(i)
        if len(vis) == 0:# all have been visited
            ans.append(res[:])
        else:
            for j in nums:
                if j in vis:# haven't been visited
                    dfs(j)
        res.pop()
        vis.add(i)
    
    for i in nums:
        dfs(i)
    return ans
```

It actually runs on different trees. Like [1,2,3] as input, [2,1,3] and [1,2,3] are on different tree for tree can only walk from root to node, if 2 is the root for the first instance, on the 2nd one, 1 can't walk to 2.

for the permutation with duplicated numbers:

```python
def permuteUnique(nums):# [1,1,2], or [1_1,1_2,2] to represent the position.
    res = []
    used = [False] * len(nums) # since the elements can be duplicated, we use "used" to determine if the element is used or not
    nums.sort()

    def backtrack(path):
        if len(path) == len(nums):
            res.append(path[:])
            return
        for i in range(len(nums)):
            if used[i]:# don't use the same element in a position twice
                continue
            if i > 0 and nums[i] == nums[i-1] and not used[i-1]:# result like [1_1,1_2,2] and [1_2,1_1,2] will never happen
                continue
            used[i] = True
            path.append(nums[i])
            backtrack(path)
            path.pop()
            used[i] = False

    backtrack([])
    return res
```

78. Subsets

```python
def subsets(nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        result = []

        def backtrack(i,path):
            result.append(path[:])

            for j in range(i + 1,len(nums)):
                path.append(nums[j])
                backtrack(j,path)
                path.pop()


        backtrack(-1,[])
        return result

print(subsets([1,2,3])) 
```
