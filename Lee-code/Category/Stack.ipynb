# 20.Valid Parenthesis

```python
def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        if len(s) == 1:
            return False
        mapping = {'(':")","{":"}","[":"]"}
        stack = []
        for char in s:
            if char in "({[":
                stack.append(char)
            else:
                if len(stack) == 0:
                    return False
                pair = stack.pop()
                if mapping[pair] != char:
                    return False
        return True if len(stack) == 0 else False
```

```python
def isValid(self, s):
        stack = []
        mapping = {')': '(', '}': '{', ']': '['}
        
        for char in s:
            if char in mapping:
                top = stack.pop() if stack else '#' # something not in the mapping
                if mapping[char] != top:
                    return False
            else:
                stack.append(char)
        
        return len(stack) == 0
```

```python
def simplifyPath(path):
        """
        :type path: str
        :rtype: str
        """
        stack = []

        for part in path.split("/"):
            if part == "..":
                if stack:
                    stack.pop()
            elif part == "." or part == "":
                continue
            else:
                stack.append(part)
        print(stack)

        return "/" + "/".join(stack)

print(simplifyPath("/home/"))
```
