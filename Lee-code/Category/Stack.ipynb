# 20.Valid Parenthesis

```python
def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """
        if len(s) == 1:
            return False
        mapping = {'(':")","{":"}","[":"]"}
        stack = []
        for char in s:
            if char in "({[":
                stack.append(char)
            else:
                if len(stack) == 0:
                    return False
                pair = stack.pop()
                if mapping[pair] != char:
                    return False
        return True if len(stack) == 0 else False
```

```python
def isValid(self, s):
        stack = []
        mapping = {')': '(', '}': '{', ']': '['}
        
        for char in s:
            if char in mapping:
                top = stack.pop() if stack else '#' # something not in the mapping
                if mapping[char] != top:
                    return False
            else:
                stack.append(char)
        
        return len(stack) == 0
```

```python
def simplifyPath(path):
        """
        :type path: str
        :rtype: str
        """
        stack = []

        for part in path.split("/"):
            if part == "..":
                if stack:
                    stack.pop()
            elif part == "." or part == "":
                continue
            else:
                stack.append(part)
        print(stack)

        return "/" + "/".join(stack)

print(simplifyPath("/home/"))
```

```python
class MinStack(object):
    
    def __init__(self):
        self.stack = []
        self.min_stack = []
        

    def push(self, value):
        """
        :type value: int
        :rtype: None
        """
        self.stack.append(value)
        if self.min_stack:
            self.min_stack.append(min(value,self.min_stack[-1]))
        else:
            self.min_stack.append(value)
        print(self.min_stack)
        

    def pop(self):
        """
        :rtype: None
        """
        self.stack.pop()
        self.min_stack.pop()
        

    def top(self):
        """
        :rtype: int
        """
        return self.stack[-1]
        

    def getMin(self):
        """
        :rtype: int
        """
        return self.min_stack[-1]
minStack = MinStack()
minStack.push(-2)
minStack.push(0) 
minStack.push(-3)
print(minStack.getMin()) # return -3   
```

150. Evaluate Reverse Polish Notation

```python
def evalRPN(tokens):
    """
    :type tokens: List[str]
    :rtype: int
    """
    stack = []
    result = 0
    for char in tokens:
        if char in "+-*/":
            first_element = stack.pop()
            second_element = stack.pop()
            if char == "+":
                result = first_element + second_element
            elif char == "-":
                result = second_element - first_element

            elif char == "*":
                result = first_element * second_element
            else:
                result = int(float(second_element) / first_element)
            stack.append(result)
        else:# number
            stack.append(int(char))
    return stack[0]
                
```

Basic Calculator

```python
class Solution(object):
    def calculate(self, s):
        """
        :type s: str
        :rtype: int
        """
        stack = []
        result,num,sign = 0,0,1
        for char in s:
            if char in "+-":
                if char == '+':
                    result += sign * num
                    num = 0
                    sign = 1
                else:
                    result += sign * num
                    num = 0
                    sign = -1
            elif char in "()":
                if char == "(":
                    stack.append(result)
                    stack.append(sign)
                    result = 0
                    sign = 1
                else:
                    result += sign * num
                    num = 0
                    result *= stack.pop()
                    result += stack.pop()
            elif char == " ":
                continue
            else:
                num = num * 10 + int(char)

        return result + sign * num     # handle last number

                
```
