# 89. Gray array formula

Recursive

```python
def grayCode(n):
    if n == 0:
        return [0]
    
    # Get gray code for n-1
    prev = grayCode(n - 1)
    
    # Mirror and add leading 1
    result = prev + [x + (1 << (n - 1)) for x in reversed(prev)]
    
    return result
```

Iterative

```python
def grayCode(n):
    result = [0]
    
    for i in range(n):
        # Add mirrored version with bit i set
        size = len(result)
        for j in range(size - 1, -1, -1):
            result.append(result[j] | (1 << i))
    
    return result
```

Mathematical Formula

```python
def grayCode(n):
    return [i ^ (i >> 1) for i in range(1 << n)]
```

```python
from collections import defaultdict

def maxPoints(points):
    if len(points) <= 2:
        return len(points)
    
    def gcd(a, b):
        """Calculate GCD manually"""
        while b:
            a, b = b, a % b
            print(f"a={a}, b={b}")
        return a
    
    max_points = 0
    
    # For each point, calculate slopes to all other points
    for i in range(len(points)):
        slopes = defaultdict(int)
        
        for j in range(i+1,len(points)):
        
            
            # Calculate slope between points[i] and points[j]
            dx = points[j][0] - points[i][0]
            dy = points[j][1] - points[i][1]
            
            # Handle vertical lines (infinite slope)
            if dx == 0:
                slope = ('inf', 0)
            else:
                # Normalize slope to avoid floating point issues
                g = gcd(abs(dx), abs(dy))
                dx //= g
                dy //= g
                
                # Ensure consistent sign (put sign in dy)
                if dx < 0:
                    dx = -dx
                    dy = -dy
                
                slope = (dy, dx)
            
            slopes[slope] += 1
        
        # The max count + 1 (for the point itself)
        if slopes:
            max_points = max(max_points, max(slopes.values()) + 1)
    
    return max_points
```

# 166: Dealing with the decimal numbers

```python
def fractionToDecimal(numerator, denominator):
        """
        :type numerator: int
        :type denominator: int
        :rtype: str
        """
        if numerator == 0:
            return "0"

        result = []

        # Handle sign
        if (numerator < 0) ^ (denominator < 0):
            result.append("-")

        numerator, denominator = abs(numerator), abs(denominator)

        # Integer part
        result.append(str(numerator // denominator))
        remainder = numerator % denominator

        if remainder == 0:
            return "".join(result)

        result.append(".")
        remainder_map = {}  # remainder -> index in result

        while remainder != 0:
            if remainder in remainder_map:
                idx = remainder_map[remainder]
                result.insert(idx, "(")
                result.append(")")
                break

            remainder_map[remainder] = len(result)
            remainder *= 10
            result.append(str(remainder // denominator))
            remainder %= denominator

        return "".join(result)
```

# 167: Every time you see the sorted array here, you eureka into binary search pattern as Pavlov's dog

Given a 1-indexed array of integers numbers that is already sorted in non-decreasing order, find two numbers such that they add up to a specific target number. Let these two numbers be numbers[index1] and numbers[index2] where 1 <= index1 < index2 <= numbers.length.

Return the indices of the two numbers, index1 and index2, added by one as an integer array [index1, index2] of length 2.

The tests are generated such that there is exactly one solution. You may not use the same element twice.

Your solution must use only constant extra space.

```python
def twoSum(numbers, target):
        """
        :type numbers: List[int]
        :type target: int
        :rtype: List[int]
        """
        nums = numbers
        left, right = 0, len(numbers) - 1

        while left < right:
            cur_sum = nums[left] + nums[right]

            if cur_sum == target:
                return [left + 1, right + 1] #idx fixing

            elif cur_sum < target:
                left += 1
            else:
                right -= 1

        return []
```

```python
def convertToTitle(columnNumber):
        """
        :type columnNumber: int
        :rtype: str
        """

        ## no CC this time



        remainder = columnNumber % 26
        result = ''

        while columnNumber != 0:
            if remainder != 0:
                result = chr(remainder + 64) + result
                columnNumber = (columnNumber - remainder) // 26
                remainder = columnNumber % 26
                print(f"remainder: {remainder}, columnNumber: {columnNumber}, result: {result}")
            else:
                columnNumber = (columnNumber - remainder) // 26
                remainder = columnNumber % 26
                print(f"remainder: {remainder}, columnNumber: {columnNumber}, result: {result}")

        return result

print(convertToTitle(701))
```
