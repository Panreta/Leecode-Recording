# Approach 1: State Machine DP (Most Intuitive):

```python
def maxProfit(prices):
    """
    :type prices: List[int]
    :rtype: int
    """
    if not prices:
        return 0
    
    # Track 4 states: after 1st buy, 1st sell, 2nd buy, 2nd sell
    buy1 = float('-inf')
    sell1 = 0
    buy2 = float('-inf')
    sell2 = 0
    
    for price in prices:
        # First transaction
        buy1 = max(buy1, -price)           # Buy at current price
        sell1 = max(sell1, buy1 + price)   # Sell after first buy
        
        # Second transaction
        buy2 = max(buy2, sell1 - price)    # Buy again after first sell
        sell2 = max(sell2, buy2 + price)   # Sell after second buy
    
    return sell2
```

# Approach 2: DP with 2D Array:


```python
def maxProfit(self, prices):
    """
    :type prices: List[int]
    :rtype: int
    """
    if not prices:
        return 0
    
    n = len(prices)
    # dp[i][j] = max profit using at most i transactions on day j
    dp = [[0] * n for _ in range(3)]  # 0, 1, or 2 transactions
    
    for t in range(1, 3):  # For 1 and 2 transactions
        max_diff = -prices[0]
        for day in range(1, n):
            # Either don't trade today, or sell today
            dp[t][day] = max(dp[t][day-1], prices[day] + max_diff)
            # Update max_diff for next iteration
            max_diff = max(max_diff, dp[t-1][day] - prices[day])
    
    return dp[2][n-1]
```

    2. how to reverse a linked list.

    3. how to merge to linked list.

so without any one of these this problem can not be done.

```python
def reorderList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: None Do not return anything, modify head in-place instead.
        """

        if not head.next:
            return 

        slow, fast = head, head
        while fast.next and fast.next.next:
            slow = slow.next
            fast = fast.next.next

        # at this moment, the slow is nearly in the half now
        # break from the half
        second = slow.next # first one of the second line
        slow.next = None
        prev = None

        #reverse the second half
        while second:
            temp = second.next
            second.next = prev
            prev = second
            second = temp

        # just merge as insersion sort
        first, second = head, prev
        while second:
            temp1,temp2 = first.next, second.next
            first.next = second
            second.next = temp1
            first, second = temp1,temp2
```

```python
def evalRPN(tokens):
        """
        :type tokens: List[str]
        :rtype: int
        """
        stack = []
        
        for token in tokens:
            if token in "+-*/":
                b = stack.pop()
                a = stack.pop()
                
                if token == '+':
                    stack.append(a + b)
                elif token == '-':
                    stack.append(a - b)
                elif token == '*':
                    stack.append(a * b)
                elif token == '/':
                    stack.append(int(a / float(b)))
                    # stack.append(a // b)
            else:
                stack.append(int(token))
        
        return stack.pop()
```

With Lambda expression

```python
def evalRPN(self, tokens):
    """
    :type tokens: List[str]
    :rtype: int
    """
    stack = []
    operations = {
        '+': lambda a, b: a + b,
        '-': lambda a, b: a - b,
        '*': lambda a, b: a * b,
        '/': lambda a, b: int(a / float(b))
    } ## hash map of operations
    
    for token in tokens:
        if token in operations:
            b = stack.pop()
            a = stack.pop()
            stack.append(operations[token](a, b))
        else:
            stack.append(int(token))
    
    return stack.pop()
```
