```python
def reverseWords(s):
    """
    :type s: str
    :rtype: str
    """
    # strip() removes leading/trailing whitespace (not trim())
    s = s.strip()
    
    # split() without arguments handles multiple spaces
    box = s.split()
    
    # Reverse the list
    box.reverse()
    
    # Join with single space
    return " ".join(box)
```

# 152:

Since the negative number flip the sign, we need to record the min also.

```python
def maxProduct(nums):
  
        # Track both max and min products ending at current position
        max_prod = nums[0]
        min_prod = nums[0]
        result = nums[0]
        
        for i in range(1, len(nums)):
            curr = nums[i]
            
            # Calculate new max and min
            # Need to check curr alone, curr*max_prod, curr*min_prod
            temp_max = max(curr, max_prod * curr, min_prod * curr)
            temp_min = min(curr, max_prod * curr, min_prod * curr)
            
            max_prod = temp_max
            min_prod = temp_min
            
      
            # Update global result
            result = max(result, max_prod)
        
        return result
```

# 153: Binary Search

```python
def findMin(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        left, right = 0, len(nums) - 1
        while left < right:
                mid = (left + right) // 2
                if nums[right] < nums[mid]:
                    left = mid + 1
                else:
                    right = mid
        return nums[left]
```

```python
def findMin(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        left, right = 0, len(nums) - 1
        while left < right:
                mid = (left + right) // 2
                if nums[right] < nums[mid]: ## left is too wide
                    left = mid + 1
                elif nums[mid] < nums[right] ## right is too wide
                    right = mid
                else: ## nums[mid] == nums[right], cannot determine, reduce right by 1
                    right -= 1
        return nums[left]
```

Change to Hexavigesimal(26-base) problem then can solve it

```python
def convertToTitle(columnNumber):

    ## no CC this time
    result = ''

    while columnNumber > 0:
        columnNumber -= 1
        remainer = columnNumber % 26
        result = chr(remainer + 65) + result
        columnNumber //= 26

    return result


```

```python
from collections import defaultdict

# Count word frequencies
words = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']
count = defaultdict(int)  # Default value is 0

for word in words:
    count[word] += 1  # No KeyError!

print(dict(count))
# {'apple': 3, 'banana': 2, 'cherry': 1}
```

# 169

you don't really care for the n/2, for some number surpass n/2 must be the largest. So just name the first one you see to be the best,then test if any number can offset this count, if can, just recount from this number

The algorithm illustrated down here is Boyer-Moore Voting Algorithm 

```python
def majorityElement(self, nums):
        cand = None
        count = 0

        for x in nums:
            if count == 0:
                cand = x
            if x == cand:
                count += 1
            else:
                count -= 1

        return cand
```

# 172. Factorial Trailing Zeroes

    Trailing zeros: zeors on the tail    
    Key insight: Trailing zeros = min(count of 2s, count of 5s) in n!
    Since 2s are always more abundant, just count 5s.

```python
mul = 1
for i in range(1,26):
    mul *= i
    print(f"i = {i}, factorial = {mul}")
    if i % 5 == 0:
        print("_"*40)
```
