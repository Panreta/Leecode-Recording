```python
def reverseWords(s):
    """
    :type s: str
    :rtype: str
    """
    # strip() removes leading/trailing whitespace (not trim())
    s = s.strip()
    
    # split() without arguments handles multiple spaces
    box = s.split()
    
    # Reverse the list
    box.reverse()
    
    # Join with single space
    return " ".join(box)
```

# 152:

Since the negative number flip the sign, we need to record the min also.

```python
def maxProduct(nums):
  
        # Track both max and min products ending at current position
        max_prod = nums[0]
        min_prod = nums[0]
        result = nums[0]
        
        for i in range(1, len(nums)):
            curr = nums[i]
            
            # Calculate new max and min
            # Need to check curr alone, curr*max_prod, curr*min_prod
            temp_max = max(curr, max_prod * curr, min_prod * curr)
            temp_min = min(curr, max_prod * curr, min_prod * curr)
            
            max_prod = temp_max
            min_prod = temp_min
            
      
            # Update global result
            result = max(result, max_prod)
        
        return result
```

```python
def findMin(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        left, right = 0, len(nums) - 1
        while left < right:
                mid = (left + right) // 2
                if nums[right] < nums[mid]:
                    left = mid + 1
                else:
                    right = mid
        return nums[left]
```

you don't really care for the n/2, for some number surpass n/2 must be the largest. So just name the first one you see to be the best,then test if any number can offset this count, if can, just recount from this number

The algorithm illustrated down here is Boyer-Moore Voting Algorithm 

```python
def majorityElement(self, nums):
        cand = None
        count = 0

        for x in nums:
            if count == 0:
                cand = x
            if x == cand:
                count += 1
            else:
                count -= 1

        return cand
```

```python
def findMinimumDays(max_products, tasks):
    products = 0
    emergency_days = 0
    
    for task in tasks:
        # Morning phase: add emergency if needed BEFORE inspection
        if task == 0 and products < 0:
            emergency_days += 1
            products = 0  # Emergency brings us to 0
        
        # Evening phase: apply task
        products += task
        
        # Check if we exceed capacity
        if products > max_products:
            return -1
    
    # Check final state
    if products < 0:
        return -1
    
    return emergency_days

max_products =  12
tasks = [-5,0,10, -13, 0]
print(findMinimumDays(max_products,tasks))
```

```python
def findMinimumDays(n,max_products,tasks):
    days = 0
    cur_products = 0
    for i in range(n):
        cur_products += tasks[i]
        if cur_products < 0:
            days += 1
            cur_products = days + 1
        elif cur_products > max_products:
            return -1
        
    return days



n = 5
max_products = 12
tasks =[-5,0, 10, -13, 0]
print(findMinimumDays(n,max_products,tasks))

max_products = 10
n = 4
tasks = [7, -2, 0, 6]
print(findMinimumDays(n,max_products,tasks))
```

# 189 Best Time to Buy and Sell Stock IV

Quite hard to understand, but we can have some basic observation:

1. if k >= n//2, it's equivalent to unlimited transactions.
2. You need DP, and the structure about DP is based on whether sell or not, so you just test which state is better.

```python
class Solution(object):
    def maxProfit(self, k, prices):
        """
        :type k: int
        :type prices: List[int]
        :rtype: int
        """
        n = len(prices)

        if k >= n//2:
            profit = 0
            for i in range(1,n):
                profit += max(0,prices[i] - prices[i-1])
            return profit

        buy = [-float('inf')] * (k + 1)
        sell = [0] * (k + 1)

        for price in prices:
            for i in range(k,0,-1):
                sell[i] = max(sell[i],buy[i] + price)
                buy[i] = max(buy[i],sell[i-1] - price)

        return sell[k]
```

# 199 consider the right side view of the tree

First Thought: 

I think I can count the depth, then always go right, if right is null but the element is not the number of depth, then go to left tree to do the same thing

Right way, BFS

```python
from collections import deque


def rightSideView(root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        if not root:
            return []
        
        result = []
        queue = deque([root])
        
        # filter the queue into result
        while queue:
            level_size = len(queue)
            
            for i in range(level_size):
                node = queue.popleft()
                
                # The last node at this level is the rightmost
                if i == level_size - 1:
                    result.append(node.val)
                
                # Add all the next level nodes
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
        
        return result
```

# 200: number of islands

Famous problem, also about BFS/DFS

```python
def numIslands(grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """
        if not grid:
            return 0
        
        rows, cols = len(grid), len(grid[0])
        islands = 0
        
        def dfs(r, c):
            # Base cases: out of bounds or water or already visited
            if (r < 0 or r >= rows or 
                c < 0 or c >= cols or 
                grid[r][c] != '1'):
                return
            
            # Mark as visited by changing '1' to '0'
            grid[r][c] = '0'
            
            # Explore all 4 directions
            dfs(r + 1, c)  # down
            dfs(r - 1, c)  # up
            dfs(r, c + 1)  # right
            dfs(r, c - 1)  # left
        
        # Scan through entire grid
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == '1':
                    islands += 1
                    dfs(r, c)  # Mark entire island as visited
        
        return islands
        
```
