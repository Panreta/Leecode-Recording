# List to array and array to List

list -> array: array(a),asarray(a)
array -> list: x.tolist()

```python
# From an existing ndarray → no copy (same object)
arr = np.array([1, 2, 3])
y = np.asarray(arr)
print(y is arr)  # True

# Compare with np.array, which always copies by default
z = np.array(arr)
print(z is arr)  # False
```

L=N1​i=1∑N​−log(pi​)
This is the final cross-entropy loss — the lower it is, the better your model is at assigning high probability to the correct class.

```python
probs = np.array([[0.7, 0.2, 0.1],   # sample 0: prob for each class
                  [0.1, 0.5, 0.4],   # sample 1
                  [0.2, 0.3, 0.5]])
X = np.array([0, 1, 2])
y = np.array([0, 1, 2])
print(probs[range(len(X)), y])
correct_logprobs = -np.log(probs[range(len(X)), y])# get [0,y[0]], [1,y[1]], [2,y[2]] from probs
data_loss = np.sum(correct_logprobs)
print(1.0/len(X) * data_loss)
```
